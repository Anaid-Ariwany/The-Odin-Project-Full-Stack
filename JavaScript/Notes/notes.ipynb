# Organizing JS Code

### Objects

#### Intro

- Objects: collections of key-value pairs. They can represent real-world entities and their properties/behaviors.

- Object literals: created using curly braces `{}`. Example:
    ```javascript
    const person = {
        name: 'Alice',
        age: 30,
        greet() {
        console.log(`Hello, my name is ${this.name}`);
        }
    };
    ```

- There are 2 ways to get information from an object:
    1. Dot notation: `person.name`
    2. Bracket notation: `person['name']`

1. `Dot Notation`:
    - Used when the property name is a valid identifier (no spaces, special characters, or reserved words).
        - Example: `person.name`
    - Can't be used on variables or expressions.
        - Example: `const prop = 'name'; person.prop` (this will not work, it will look for a property named 'prop' instead of 'name')
    
2. `Bracket Notation`:
    - Used when the property name is not a valid identifier or is dynamic.
        - Example: `person['name']` or `person["first name"]`
    - Can be used with variables or expressions.
        - Example: `const prop = 'name'; person[prop]` (this will work, it will look for a property named 'name')

##### This Keyword

- `this` refers to the current object context. It allows access to the object's properties and methods from within its own methods.
    - Example:
    ```javascript
        const person = {
            name: 'Alice',
            greet() {
                console.log(`Hello, my name is ${this.name}`);
            }
        };
        person.greet(); // Output: Hello, my name is Alice
     ```
- The value of `this` can change based on how a function is called. In a method, `this` refers to the object. In a regular function, `this` refers to the global object (or `undefined` in strict mode).
    - Example:
    ```javascript   
        function showThis() {
            console.log(this);
        }
    showThis(); // In non-strict mode, this will log the global object (window in browsers). In strict mode, it will log undefined.
```



- In simpler terms, `this` is a special keyword that allows you to refer to the current object within its own methods. It helps you access the object's properties and methods without having to hardcode the object name, making your code more flexible and reusable.

- Example:
    ```javascript
        const car = {
            make: "Volkswagen",
            model: "Golf",
            year: 2026,
            color: "blue",
            priceUSD: 40000,
        };

        function displayCarInfo() {
        console.log(`This car is a ${this.year} ${this.make} ${this.model} in ${this.color} color, priced at $${this.priceUSD}.`);
        }
        
        car.displayInfo = displayCarInfo;
        car.displayInfo(); // Output: This car is a 2026 Volkswagen Golf in blue color, priced at $40000.
    ```

#### Object Constructors

##### Intro

- **_Object Constructors_** are functions that create and initialize (construct or in simpler terms, build) objects. They are a way to create multiple instances _(instance is a single object created from a constructor; a constructor is a function used to create and initialize objects)_ of an object with the same structure but different values.

- Example: 
    ```javascript
        function Car(make, model, year, color, priceUSD) {
            this.make = make;
            this.model = model;
            this.year = year;
            this.color = color;
            this.priceUSD = priceUSD;
        }

        const car1 = new Car("Volkswagen", "Golf", 2026, "blue", 40000);
        const car2 = new Car("Toyota", "Camry", 2025, "red", 35000);

        console.log(car1); // Output: Car { make: 'Volkswagen', model: 'Golf', year: 2026, color: 'blue', priceUSD: 40000 }
        console.log(car2); // Output: Car { make: 'Toyota', model: 'Camry', year: 2025, color: 'red', priceUSD: 35000 }
    ```
    - Explanation:
        - The `Car` function is a constructor that takes parameters to initialize the properties of the car object.
        - The `new` keyword is used to create new instances of the `Car` object, which allows us to create multiple cars with different properties while sharing the same structure defined by the constructor.
        - Each instance (car1, car2) has its own set of properties based on the values passed to the constructor.
        - This approach promotes code reusability and organization when creating multiple objects with similar characteristics.
        - The `this.name = name;` syntax inside the constructor assigns the value of the `name` parameter to the `name` property of the object being created.

##### The Prototype

- **_Prototypes_** are a fundamental concept in JavaScript that allows objects to inherit properties and methods from other objects. Every JavaScript object has a prototype, which is another object that it can delegate property and method lookups to.

- The prototype is a powerful mechanism for sharing behavior among objects. When you access a property or method on an object, JavaScript first looks for it on the object itself. If it doesn't find it there, it looks up the prototype chain until it finds the property or reaches the end of the chain (null).

- The syntax for adding properties or methods to a constructor's prototype is:
    ```javascript
        ConstructorName.prototype.methodName = function() {
            // method body
        };
    ```
- Example:
    ```javascript
        function Car(make, model, year) {
            this.make = make;
            this.model = model;
            this.year = year;
        }

        Car.prototype.getCarInfo = function() {
            return `${this.year} ${this.make} ${this.model}`;
        };

        const car1 = new Car("Volkswagen", "Golf", 2026);
        console.log(car1.getCarInfo()); // Output: 2026 Volkswagen Golf
    ```
- Explanation:
    - In this example, we define a constructor function `Car` that initializes the properties of a car object.
    - We then add a method `getCarInfo` to the `Car` prototype, which allows all instances of `Car` to access this method without having to define it on each instance.
    - When we create an instance `car1` and call `car1.getCarInfo()`, JavaScript looks for the `getCarInfo` method on `car1`. Since it doesn't find it there, it looks up the prototype chain and finds it on `Car.prototype`, allowing us to call the method successfully.

1. _All Objects in JS have a `[[Prototype]]`:_
    - This is an internal property that points to the prototype object from which the object inherits properties and methods. It is not directly accessible in code, but it can be accessed using `Object.getPrototypeOf()` or the `__proto__` property (though the latter is deprecated).
    - Example:
    ```javascript
        const car1 = new Car("Volkswagen", "Golf", 2026);
        console.log(Object.getPrototypeOf(car1)); // Output: Car { getCarInfo: [Function] }
    ```
    - In simpler terms, prototypes allow us to define methods and properties that can be shared across all instances of a constructor function, promoting code reusability and efficient memory usage.


2. _The `[[Prototype]]` is another object:_
    - The prototype itself is an object that can have its own properties and methods. This means that prototypes can also inherit from other prototypes, creating a chain of inheritance.
    - Example:
    ```javascript
        function Vehicle() {
            this.type = "vehicle";
        }

        Vehicle.prototype.getType = function() {
            return this.type;
        };

        function Car(make, model, year) {
            Vehicle.call(this); // Call the Vehicle constructor to initialize the type property
            this.make = make;
            this.model = model;
            this.year = year;
        }

        Car.prototype = Object.create(Vehicle.prototype); // Set Car's prototype to be an object created from Vehicle's prototype
        Car.prototype.constructor = Car; // Set the constructor property back to Car

        const car1 = new Car("Volkswagen", "Golf", 2026);
        console.log(car1.getType()); // Output: vehicle
    ```
    - In this example, we have a `Vehicle` constructor with a method `getType`. The `Car` constructor inherits from `Vehicle` by setting its prototype to an object created from `Vehicle.prototype`. This allows instances of `Car` to access the `getType` method defined in `Vehicle`, demonstrating how prototypes can create a chain of inheritance.


3. _That the original object inherits from, has access to all of its `[[Prototype]]` properties and methods:_
    - When an object inherits from a prototype, it has access to all the properties and methods defined on that prototype. This means that if a property or method is not found on the object itself, JavaScript will look up the prototype chain to find it.
    - Example:
    ```javascript
        function Vehicle() {
            this.type = "vehicle";
        }

        Vehicle.prototype.getType = function() {
            return this.type;
        };

        function Car(make, model, year) {
            Vehicle.call(this); // Call the Vehicle constructor to initialize the type property
            this.make = make;
            this.model = model;
            this.year = year;
        }

        Car.prototype = Object.create(Vehicle.prototype); // Set Car's prototype to be an object created from Vehicle's prototype
        Car.prototype.constructor = Car; // Set the constructor property back to Car

        const car1 = new Car("Volkswagen", "Golf", 2026);
        console.log(car1.getType()); // Output: vehicle
    ```
    - In this example, `car1` is an instance of `Car`, which inherits from `Vehicle`. Even though `car1` does not have its own `getType` method, it can access it through the prototype chain, demonstrating that the original object (car1) has access to all of its `[[Prototype]]` properties and methods.

- Why use prototypes?
    1. We can define properties and functions common among all objects on a prototype to save memory. Defining every property and function takes up a lot of memory, especially if you have a lot of common properties and functions, and a lot of created objects! Defining them on a centralized, shared object which the objects have access to, thus saves memory.

    2. The second reason is the name of this section, `Prototypal Inheritance`, which we’ve referred to in passing earlier, in the introduction to the Prototype. In recap, we can say that the `player1` and `player2` objects inherit from the Player.prototype object, which allows them to access functions like `.sayHello`.

### Factory Functions & the Module Pattern

#### Variable Scope

- **Scoping:** the accessibility of variables and functions in different parts of the code. JavaScript has function scope and block scope.
    - **Function scope:** variables declared within a function are only accessible within that function.
    - **Block scope:** variables declared with `let` or `const` within a block (e.g., inside an `if` statement or a loop) are only accessible within that block.
- Example:
    ```javascript
        function exampleFunction() {
            var functionScopedVar = "I am function scoped";
            if (true) {
                let blockScopedVar = "I am block scoped";
                console.log(functionScopedVar); // Output: I am function scoped
                console.log(blockScopedVar); // Output: I am block scoped
            }
            console.log(functionScopedVar); // Output: I am function scoped
            console.log(blockScopedVar); // Error: blockScopedVar is not defined
        }
        exampleFunction();
    ```

- There are 2 types of scope:
    1. **Global Scope:** variables declared outside of any function or block are in the global scope and can be accessed from anywhere in the code.
        - Example: 
        ```javascript
            var globalVar = "I am global";
            function exampleFunction() {
                console.log(globalVar); // Output: I am global
            }
            exampleFunction();
        ```

    2. **Local Scope:** variables declared within a function or block are in the local scope and can only be accessed within that function or block.
        - Example:
        ```javascript
            function exampleFunction() {
                var localVar = "I am local";
                console.log(localVar); // Output: I am local
            }
            exampleFunction();
            console.log(localVar); // Error: localVar is not defined
        ```

#### Closures

- **Closures** are a fundamental concept in JavaScript that allows a function to access variables from its outer (enclosing) scope even after the outer function has finished executing. A closure is created when a function is defined inside another function and retains access to the outer function's variables.
- Example:
    ```javascript
        function outerFunction() {
            let outerVariable = "I am from the outer function";
            
            function innerFunction() {
                console.log(outerVariable); // Output: I am from the outer function
            }
            
            return innerFunction;
        }

        const closureExample = outerFunction();
        closureExample(); // Output: I am from the outer function
    ```
    - Explanation:
        - In this example, `outerFunction` defines a variable `outerVariable` and an inner function `innerFunction` that accesses `outerVariable`.
        - When `outerFunction` is called, it returns `innerFunction`, which retains access to `outerVariable` even after `outerFunction` has finished executing. This is the essence of a closure: the inner function "closes over" the variables of the outer function, allowing it to access and manipulate those variables even when the outer function is no longer in scope.

- **Surrounding State / Lexical Environment:** the variables and functions that are in scope at the time a closure is created. The surrounding state is what allows closures to access and manipulate variables from their outer scope.
- Example:
    ```javascript
        function createCounter() {
            let count = 0; // This is the surrounding state
            
            return function() { // This inner function is the closure
                count++; // Accessing and modifying the surrounding state
                console.log(count);
            };
        }

        const counter = createCounter();
        counter(); // Output: 1
        counter(); // Output: 2
        counter(); // Output: 3
    ```
    - Explanation:
        - In this example, `createCounter` initializes a variable `count` and returns an inner function that increments and logs `count`.
        - The inner function is a closure that has access to the surrounding state (the `count` variable) even after `createCounter` has finished executing. Each time we call `counter()`, it updates and logs the current value of `count`, demonstrating how closures can maintain and manipulate state across multiple invocations.

#### Factory Functions

- **Factory Functions** are functions that create and return new objects. They are a way to create multiple instances of an object with the same structure but different values without using classes or constructors.
- Example:
    ```javascript
        function createCar(make, model, year) {
            return {
                make: make,
                model: model,
                year: year,
                getCarInfo() {
                    return `${this.year} ${this.make} ${this.model}`;
                }
            };
        }

        const car1 = createCar("Volkswagen", "Golf", 2026);
        const car2 = createCar("Toyota", "Camry", 2025);

        console.log(car1.getCarInfo()); // Output: 2026 Volkswagen Golf
        console.log(car2.getCarInfo()); // Output: 2025 Toyota Camry
    ```
    - Explanation:
        - In this example, `createCar` is a factory function that takes parameters to create and return a new car object with properties and a method.
        - Each time we call `createCar`, it returns a new object with the specified properties and the `getCarInfo` method. This allows us to create multiple car objects with different values while sharing the same structure defined by the factory function.

- Advantages of using factory functions over constructors:
    1. Factory functions can return any type of object, not just instances of a specific constructor. This allows for more flexibility in the types of objects that can be created.

    2. Factory functions can have private variables and methods, which are not accessible from outside the function. This allows for better encapsulation and data hiding.
    
    3. Factory functions do not require the use of the `new` keyword, which can help prevent errors and improve readability.

#### Private Variables & Functions

- **Private Variables & Functions** are variables and functions that are not accessible from outside the function or object they are defined in. They are used to encapsulate data and behavior, providing a way to hide implementation details and protect the internal state of an object.

- Example:
    ```javascript
        function createCounter() {
            let count = 0; // This is a private variable
            
            function increment() { // This is a private function
                count++;
                console.log(count);
            }
            
            return {
                increment // Exposing the increment function to the outside
            };
        }

        const counter = createCounter();
        counter.increment(); // Output: 1
        counter.increment(); // Output: 2
        console.log(counter.count); // Output: undefined (count is private)
    ```
    - Explanation:
        - In this example, `count` is a private variable that cannot be accessed directly from outside the `createCounter` function. The `increment` function is also private and can only be called through the object returned by `createCounter`.
        - By using closures, we can create private variables and functions that are encapsulated within the factory function, allowing us to control access to them and protect the internal state of our objects.

#### IIFEs & the Module Pattern

- **Module Pattern** is a design pattern used to create modules in JavaScript. It allows us to encapsulate code and create private variables and functions while exposing only the necessary parts of the module to the outside world. This is often achieved using an Immediately Invoked Function Expression (IIFE).

- **IIEFs (Immediately Invoked Function Expressions)** are functions that are defined and immediately executed. They are often used to create a new scope and avoid polluting the global namespace.
- Example:
    ```javascript
        (function() {
            const privateVariable = "I am private";
            console.log(privateVariable);
        })();
    ```
    - Explanation:
        - In this example, we define an anonymous function and immediately invoke it. This creates a new scope for the `privateVariable`, which is not accessible from outside the function. This pattern is useful for encapsulating code and preventing variables from leaking into the global scope.

### Classes

#### Intro

- JS does not have traditional classes like other programming languages, but it has a class syntax that provides a more familiar way to create objects and handle inheritance. The class syntax is syntactic sugar over the existing prototype-based inheritance in JavaScript, making it easier to write and understand object-oriented code.
- Example:
    ```javascript
        class Car {
            constructor(make, model, year) {
                this.make = make;
                this.model = model;
                this.year = year;
            }

            getCarInfo() {
                return `${this.year} ${this.make} ${this.model}`;
            }
        }

        const car1 = new Car("Volkswagen", "Golf", 2026);
        console.log(car1.getCarInfo()); // Output: 2026 Volkswagen Golf
    ```
    - Explanation:
        - In this example, we define a class `Car` with a constructor that initializes the properties of the car object and a method `getCarInfo` that returns a string with the car's information.
        - We create an instance of the `Car` class using the `new` keyword, which allows us to call the constructor and initialize the object. The class syntax provides a clearer and more concise way to define objects and their behavior compared to using constructor functions and prototypes directly.

- **_Basic class syntax:_**
    ```javascript
        class ClassName {
            constructor(parameters) {
                // initialization code
            }

            methodName() {
                // method code
            }
        }
    ```
- The `constructor` method is a special method for creating and initializing an object created with a class. It is called automatically when a new instance of the class is created using the `new` keyword. The `constructor` can take parameters to initialize the properties of the object.

- _Differences between an object constructor and a class:_
    1. Syntax: Classes use a more concise and familiar syntax compared to constructor functions, making it easier to read and write object-oriented code.

    2. Inheritance: Classes provide a built-in way to handle inheritance using the `extends` keyword, while constructor functions require manual setup of the prototype chain for inheritance.

    3. Static Methods: Classes allow you to define static methods that are called on the class itself, rather than on instances of the class. Constructor functions do not have this feature.

    4. Hoisting: Class declarations are not hoisted, meaning you cannot use a class before it is declared in the code. Constructor functions, on the other hand, are hoisted and can be used before their declaration.

- **Getters and Setters** are special methods in classes that allow you to define how to access and modify the properties of an object. 
    1. Getters are used to retrieve the value of a property, while 
    
    2. setters are used to set the value of a property.

- Example:
    ```javascript
        class Car {
            constructor(make, model, year) {
                this.make = make;
                this.model = model;
                this.year = year;
            }

            get carInfo() { // Getter
                return `${this.year} ${this.make} ${this.model}`;
            }

            set updateYear(newYear) { // Setter
                this.year = newYear;
            }
        }

        const car1 = new Car("Volkswagen", "Golf", 2026);
        console.log(car1.carInfo); // Output: 2026 Volkswagen Golf

        car1.updateYear = 2027; // Using the setter to update the year
        console.log(car1.carInfo); // Output: 2027 Volkswagen Golf
    ```
    - Explanation:
        - In this example, we define a getter `carInfo` that returns a string with the car's information and a setter `updateYear` that allows us to update the year of the car.
        - We can access the getter as if it were a property (`car1.carInfo`) and use the setter to update the year by assigning a new value to it (`car1.updateYear = 2027`). This provides a clean and intuitive way to manage object properties while encapsulating the logic for getting and setting values.

#### Inheritance with Classes

- **_Inheritance_** is a fundamental concept in object-oriented programming that allows a new class (called a `subclass` or `child class`) to inherit properties and methods from an existing class (called a `superclass` or `parent class`). In JavaScript, we can achieve inheritance using the `extends` keyword in class syntax.

- Example:
    ```javascript
        class Vehicle {
            constructor(type) {
                this.type = type;
            }

            getType() {
                return this.type;
            }
        }

        class Car extends Vehicle {
            constructor(make, model, year) {
                super("car"); // Call the parent class constructor
                this.make = make;
                this.model = model;
                this.year = year;
            }
        }
        const car1 = new Car("Volkswagen", "Golf", 2026);
        console.log(car1.getType()); // Output: car (inherited from Vehicle)
    ```
    - Explanation:
        - In this example, we have a `Vehicle` class with a method `getType`. The `Car` class extends `Vehicle`, which means it inherits the properties and methods of `Vehicle`.
        - The `super("car")` call in the `Car` constructor is used to call the constructor of the parent class (`Vehicle`) and initialize the `type` property.
        - When we create an instance of `Car` and call `car1.getType()`, it successfully accesses the inherited method from the `Vehicle` class, demonstrating how inheritance works with classes in JavaScript.

#### Private Class Fields & Methods

- A private class field is a property that is only accessible within the class it is defined in. It is denoted by a `#` prefix before the field name. Private fields cannot be accessed or modified from outside the class, providing encapsulation and data hiding.

- Example:
    ```javascript
        class Car {
            #make; // Private field
            #model; // Private field

            constructor(make, model) {
                this.#make = make;
                this.#model = model;
            }

            getCarInfo() {
                return `${this.#make} ${this.#model}`;
            }
        }

        const car1 = new Car("Volkswagen", "Golf");
        console.log(car1.getCarInfo()); // Output: Volkswagen Golf
        console.log(car1.#make); // Error: Private field '#make' must be declared in an enclosing class
    ```
    - Explanation:
        - In this example, `#make` and `#model` are private fields that cannot be accessed from outside the `Car` class. The `getCarInfo` method can access these private fields because it is defined within the class.
        - Attempting to access `car1.#make` from outside the class results in an error, demonstrating that the private fields are truly private and cannot be accessed or modified from outside the class. This feature allows for better encapsulation and data hiding in JavaScript classes.

- Private methods can also be defined in a similar way using the `#` prefix. They are only accessible within the class and cannot be called from outside the class.

- Example:
    ```javascript
        class Car {
            #make; // Private field
            #model; // Private field

            constructor(make, model) {
                this.#make = make;
                this.#model = model;
            }

            #getCarInfo() { // Private method
                return `${this.#make} ${this.#model}`;
            }

            publicGetCarInfo() { // Public method that calls the private method
                return this.#getCarInfo();
            }
        }

        const car1 = new Car("Volkswagen", "Golf");
        console.log(car1.publicGetCarInfo()); // Output: Volkswagen Golf
        console.log(car1.#getCarInfo()); // Error: Private method '#getCarInfo' must be declared in an enclosing class
    ```
    - Explanation:
        - In this example, `#getCarInfo` is a private method that can only be called from within the `Car` class. The `publicGetCarInfo` method is a public method that calls the private method to retrieve the car information.
        - Attempting to call `car1.#getCarInfo()` from outside the class results in an error, demonstrating that private methods are also encapsulated and cannot be accessed from outside the class. This allows for better control over the internal behavior of the class while still providing a public interface for interacting with it.

#### Static Properties & Methods

- Static properties and methods are associated with the class itself rather than instances of the class. They can be accessed directly on the class without needing to create an instance. Static members are defined using the `static` keyword.

- Example:
    ```javascript
        class Car {
            static numberOfWheels = 4; // Static property

            constructor(make, model) {
                this.make = make;
                this.model = model;
            }

            static getNumberOfWheels() { // Static method
                return Car.numberOfWheels;
            }
        }

        console.log(Car.numberOfWheels); // Output: 4 (accessing static property)
        console.log(Car.getNumberOfWheels()); // Output: 4 (calling static method)
    ```
    - Explanation:
        - In this example, `numberOfWheels` is a static property that belongs to the `Car` class, and `getNumberOfWheels` is a static method that returns the value of the static property.
        - We can access the static property and call the static method directly on the `Car` class without creating an instance. This is useful for properties and methods that are relevant to the class as a whole rather than individual instances.

### Modules

- **Modules:** are a way to organize and encapsulate code in JavaScript. They allow you to break your code into smaller, reusable pieces, and control the visibility of variables and functions. Modules can be implemented using various patterns, such as the Module Pattern or ES6 modules. In simpler terms, modules help you keep your code organized and prevent naming conflicts by encapsulating functionality and exposing only what is necessary to the outside world.

#### Import & Export

- In Modules, we can use `export` to make variables, functions, or classes available for use in other files, and `import` to bring those exported members into another file.
- Example:
    ```javascript
        // In mathUtils.js
        export function add(a, b) {
            return a + b;
        }

        export const PI = 3.14159;

        // In main.js
        import { add, PI } from './mathUtils.js';

        console.log(add(2, 3)); // Output: 5
        console.log(PI); // Output: 3.14159
    ```
    - Explanation:
        - In this example, we have a module `mathUtils.js` that exports a function `add` and a constant `PI`. In another file `main.js`, we import these members using the `import` statement.
        - This allows us to keep our code organized by separating functionality into different files while still being able to use the exported members across our application.

- There are 2 types of importing and exporting:
    1. **Named Exports/Imports:** You can export multiple variables, functions, or classes from a module using named exports. When importing, you need to use the same names as the exported members. There are 2 ways to use `named` exports:
        * Exporting/Importing as you declare:

        ```javascript
            export function add(a, b) {
                return a + b;
            }

            export const PI = 3.14159;
        ```

        ```javascript
        //Importing
            import { add, PI } from './mathUtils.js';
        ```


        * Exporting/Importing after you declare:
        ```javascript
            function add(a, b) {
                return a + b;
            }

            const PI = 3.14159;

            export { add, PI };
        ```

        ```javascript
        //importing
            import { add, PI } from './mathUtils.js';
        ```



    

2. **Default Exports/Imports:** A module can have **_only one_** default export, which can be a variable, function, or class. When importing a default export, you can choose any name for it.
    - Example:
        ```javascript
            // In mathUtils.js
            export default function subtract(a, b) {
                return a - b;
            }

            // In main.js
            import subtract from './mathUtils.js';

            console.log(subtract(5, 2)); // Output: 3
        ```
        - Explanation:
            - In this example, we have a default export of the `subtract` function in `mathUtils.js`. When we import it in `main.js`, we can choose any name for the imported function (in this case, we also called it `subtract` for clarity).
            - Default exports are useful when you want to export a single main functionality from a module, while named exports are better for exporting multiple related functionalities.

    - Another example of default export:
        ```javascript
            // In car.js
            export default class Car {
                constructor(make, model) {
                    this.make = make;
                    this.model = model;
                }
            }

            // In main.js
            import Car from './car.js';

            const myCar = new Car("Volkswagen", "Golf");
            console.log(myCar); // Output: Car { make: 'Volkswagen', model: 'Golf' }
        ```
        - Explanation:
            - In this example, we have a default export of the `Car` class in `car.js`. When we import it in `main.js`, we can create an instance of the `Car` class and use it as needed. This demonstrates how default exports can be used to export classes or other main functionalities from a module.

- You can combine named and default exports in the same module, but you need to be careful with the syntax when importing them.
    - Example:
        ```javascript
            // In mathUtils.js
            export function add(a, b) {
                return a + b;
            }

            export const PI = 3.14159;

            export default function subtract(a, b) {
                return a - b;
            }

            // In main.js
            import subtract, { add, PI } from './mathUtils.js';

            console.log(add(2, 3)); // Output: 5
            console.log(PI); // Output: 3.14159
            console.log(subtract(5, 2)); // Output: 3
        ```
        - Explanation:
            - In this example, we have both named exports (`add` and `PI`) and a default export (`subtract`) in the `mathUtils.js` module. When importing, we first import the default export (which can be named anything) and then the named exports using their specific names. This allows us to use all the exported members from the module in our `main.js` file.

#### Entry Points

- When we use ESM (ECMAScript Modules) instead of adding every JavaScript file to our HTML in order, we only need to link a single file - the **_entry point_**. 
- The `entry point` is the main JavaScript file that serves as the starting point for our application. It is responsible for importing other modules and initializing the application. By using an entry point, we can manage our code more efficiently and keep our HTML cleaner, as we only need to include one script tag that points to the entry point file.

- We mostly use the `index.js` file as the entry point, but it can be named anything. The important thing is that it serves as the main file that imports other modules (dependent on other files) and starts the application.

- Declaration: 
    ```html
        <script type="module" src="index.js"></script>
    ```

#### Common JS (CJS)

- CJS (CommonJS) is a module system used in Node.js for organizing and structuring code. It allows developers to create reusable modules and manage dependencies in a modular way. In CommonJS, we use `require` to import modules and `module.exports` to export them.
- Example:
    ```javascript
        // In mathUtils.js
        function add(a, b) {
            return a + b;
        }

        const PI = 3.14159;

        module.exports = { add, PI };

        // In main.js
        const { add, PI } = require('./mathUtils.js');

        console.log(add(2, 3)); // Output: 5
        console.log(PI); // Output: 3.14159
    ```
    - Explanation:
        - In this example, we define a module `mathUtils.js` that exports a function `add` and a constant `PI` using `module.exports`. In another file `main.js`, we import these members using the `require` function.
        - CommonJS is primarily used in Node.js environments and is not natively supported in browsers. However, tools like Browserify or Webpack can be used to bundle CommonJS modules for use in the browser.

### npm

- **npm:** is a package manager for JavaScript that allows developers to easily share and reuse code. npm provides a vast ecosystem of open-source libraries and tools that can be easily installed and managed in your projects. With npm, you can install packages from the npm registry, manage dependencies, and run scripts to automate tasks in your development workflow.

- In simpler terms, `npm` is a tool that helps you find and use code written by other developers, making it easier to build your applications without having to reinvent the wheel.

- To use npm, you need to have `Node.js` installed on your machine, as npm comes bundled with Node.js. Once you have Node.js installed, you can use npm to manage your project dependencies and run scripts.

#### package.json

- `package.json` is a file that contains metadata about your project and its dependencies. It is used by npm to manage your project's dependencies, scripts, and other configurations. The `package.json` file is typically located in the root directory of your project. It includes information such as the project name, version, description, author, license, and most importantly, the dependencies and scripts.

- It’s a JSON _(JavaScript Object Notation)_ file containing information about our project, such as its name or any dependencies and their version numbers. npm can read this file and do things such as install all of the listed dependencies with the correct versions, and running commands that you’ve set as an npm script

- Example: 
    ```json
        {
            "name": "my-project",
            "version": "1.0.0",
            "description": "A sample project using npm",
            "main": "index.js",
            "scripts": {
                "start": "node index.js",
                "test": "echo \"Error: no test specified\" && exit 1"
            },
            "author": "Your Name",
            "license": "ISC",
            "dependencies": {
                "express": "^4.17.1"
            }
        }
    ```
    - Explanation:
        - In this example, we have a `package.json` file that defines a project with the name `my-project`, version `1.0.0`, and a description. It specifies `index.js` as the main entry point of the application.
        
        - The `scripts` section defines two scripts: `start`, which runs the application using Node.js, and `test`, which is a placeholder for running tests.

        - The `dependencies` section lists the packages that the project depends on, in this case, the `express` package with a version of `^4.17.1`. This means that when we run `npm install`, npm will install the specified version of Express and its dependencies.

        - The `license` field specifies the license under which the project is released, and the `author` field provides information about the creator of the project. Overall, the `package.json` file serves as a central configuration file for managing the project's dependencies and scripts.

- Read more here: [the package.json file](https://docs.npmjs.com/creating-a-package-json-file)

##### Installing packages with npm

- npm is used to install packages _(dependencies, or in simpler terms, libraries)_ from the npm registry, which is a large collection of open-source libraries and tools. You can install packages globally or locally in your project.

- To install a package locally in your project, you can use the following command:
    ```bash
        npm install package-name
    ```
    - This will install the specified package and add it to the `dependencies` section of your `package.json` file.

- Read more here: [installing packages with npm](https://docs.npmjs.com/downloading-and-installing-packages-locally) 

### Webpack

### JSON

### OOP Principles

# JS In the Real World

# Asynchronous JS & APIs

# Testing JS

# A Bit of Computer Science

# Intermediate Git
