# .gitignore

- When working with packages that are installed with npm, you donâ€™t need to track the contents of `node_modules` with git, nor push those files to GitHub. As we learned in the npm lesson, the `package.json` file contains all the dependency information, so that anyone can clone your project and install them on their machine with `npm install`.

- You can make a `.gitignore` file in the root of the project, and by writing file or directory names in it, you can tell git what things you donâ€™t want to track. Itâ€™s customary to add `node_modules` to `.gitignore`, since it can get really big. Similarly, `dist` is often ignored as it can be generated when someone runs the command to bundle/build the application.

- When creating a new repo on GitHub, there is an option to specify a `.gitignore` template. There are many templates out there that include common files and directories that are not typically tracked based on the type of project or language used. For JavaScript projects, there is a `node` template that includes `node_modules` and `dist`.

# Organizing JS Code

### Objects

#### Intro

- Objects: collections of key-value pairs. They can represent real-world entities and their properties/behaviors.

- Object literals: created using curly braces `{}`. Example:
    ```javascript
    const person = {
        name: 'Alice',
        age: 30,
        greet() {
        console.log(`Hello, my name is ${this.name}`);
        }
    };
    ```

- There are 2 ways to get information from an object:
    1. Dot notation: `person.name`
    2. Bracket notation: `person['name']`

1. `Dot Notation`:
    - Used when the property name is a valid identifier (no spaces, special characters, or reserved words).
        - Example: `person.name`
    - Can't be used on variables or expressions.
        - Example: `const prop = 'name'; person.prop` (this will not work, it will look for a property named 'prop' instead of 'name')
    
2. `Bracket Notation`:
    - Used when the property name is not a valid identifier or is dynamic.
        - Example: `person['name']` or `person["first name"]`
    - Can be used with variables or expressions.
        - Example: `const prop = 'name'; person[prop]` (this will work, it will look for a property named 'name')

##### This Keyword

- `this` refers to the current object context. It allows access to the object's properties and methods from within its own methods.
    - Example:
    ```javascript
        const person = {
            name: 'Alice',
            greet() {
                console.log(`Hello, my name is ${this.name}`);
            }
        };
        person.greet(); // Output: Hello, my name is Alice
     ```
- The value of `this` can change based on how a function is called. In a method, `this` refers to the object. In a regular function, `this` refers to the global object (or `undefined` in strict mode).
    - Example:
    ```javascript   
        function showThis() {
            console.log(this);
        }
    showThis(); // In non-strict mode, this will log the global object (window in browsers). In strict mode, it will log undefined.
```



- In simpler terms, `this` is a special keyword that allows you to refer to the current object within its own methods. It helps you access the object's properties and methods without having to hardcode the object name, making your code more flexible and reusable.

- Example:
    ```javascript
        const car = {
            make: "Volkswagen",
            model: "Golf",
            year: 2026,
            color: "blue",
            priceUSD: 40000,
        };

        function displayCarInfo() {
        console.log(`This car is a ${this.year} ${this.make} ${this.model} in ${this.color} color, priced at $${this.priceUSD}.`);
        }
        
        car.displayInfo = displayCarInfo;
        car.displayInfo(); // Output: This car is a 2026 Volkswagen Golf in blue color, priced at $40000.
    ```

#### Object Constructors

##### Intro

- **_Object Constructors_** are functions that create and initialize (construct or in simpler terms, build) objects. They are a way to create multiple instances _(instance is a single object created from a constructor; a constructor is a function used to create and initialize objects)_ of an object with the same structure but different values.

- Example: 
    ```javascript
        function Car(make, model, year, color, priceUSD) {
            this.make = make;
            this.model = model;
            this.year = year;
            this.color = color;
            this.priceUSD = priceUSD;
        }

        const car1 = new Car("Volkswagen", "Golf", 2026, "blue", 40000);
        const car2 = new Car("Toyota", "Camry", 2025, "red", 35000);

        console.log(car1); // Output: Car { make: 'Volkswagen', model: 'Golf', year: 2026, color: 'blue', priceUSD: 40000 }
        console.log(car2); // Output: Car { make: 'Toyota', model: 'Camry', year: 2025, color: 'red', priceUSD: 35000 }
    ```
    - Explanation:
        - The `Car` function is a constructor that takes parameters to initialize the properties of the car object.
        - The `new` keyword is used to create new instances of the `Car` object, which allows us to create multiple cars with different properties while sharing the same structure defined by the constructor.
        - Each instance (car1, car2) has its own set of properties based on the values passed to the constructor.
        - This approach promotes code reusability and organization when creating multiple objects with similar characteristics.
        - The `this.name = name;` syntax inside the constructor assigns the value of the `name` parameter to the `name` property of the object being created.

##### The Prototype

- **_Prototypes_** are a fundamental concept in JavaScript that allows objects to inherit properties and methods from other objects. Every JavaScript object has a prototype, which is another object that it can delegate property and method lookups to.

- The prototype is a powerful mechanism for sharing behavior among objects. When you access a property or method on an object, JavaScript first looks for it on the object itself. If it doesn't find it there, it looks up the prototype chain until it finds the property or reaches the end of the chain (null).

- The syntax for adding properties or methods to a constructor's prototype is:
    ```javascript
        ConstructorName.prototype.methodName = function() {
            // method body
        };
    ```
- Example:
    ```javascript
        function Car(make, model, year) {
            this.make = make;
            this.model = model;
            this.year = year;
        }

        Car.prototype.getCarInfo = function() {
            return `${this.year} ${this.make} ${this.model}`;
        };

        const car1 = new Car("Volkswagen", "Golf", 2026);
        console.log(car1.getCarInfo()); // Output: 2026 Volkswagen Golf
    ```
- Explanation:
    - In this example, we define a constructor function `Car` that initializes the properties of a car object.
    - We then add a method `getCarInfo` to the `Car` prototype, which allows all instances of `Car` to access this method without having to define it on each instance.
    - When we create an instance `car1` and call `car1.getCarInfo()`, JavaScript looks for the `getCarInfo` method on `car1`. Since it doesn't find it there, it looks up the prototype chain and finds it on `Car.prototype`, allowing us to call the method successfully.

1. _All Objects in JS have a `[[Prototype]]`:_
    - This is an internal property that points to the prototype object from which the object inherits properties and methods. It is not directly accessible in code, but it can be accessed using `Object.getPrototypeOf()` or the `__proto__` property (though the latter is deprecated).
    - Example:
    ```javascript
        const car1 = new Car("Volkswagen", "Golf", 2026);
        console.log(Object.getPrototypeOf(car1)); // Output: Car { getCarInfo: [Function] }
    ```
    - In simpler terms, prototypes allow us to define methods and properties that can be shared across all instances of a constructor function, promoting code reusability and efficient memory usage.


2. _The `[[Prototype]]` is another object:_
    - The prototype itself is an object that can have its own properties and methods. This means that prototypes can also inherit from other prototypes, creating a chain of inheritance.
    - Example:
    ```javascript
        function Vehicle() {
            this.type = "vehicle";
        }

        Vehicle.prototype.getType = function() {
            return this.type;
        };

        function Car(make, model, year) {
            Vehicle.call(this); // Call the Vehicle constructor to initialize the type property
            this.make = make;
            this.model = model;
            this.year = year;
        }

        Car.prototype = Object.create(Vehicle.prototype); // Set Car's prototype to be an object created from Vehicle's prototype
        Car.prototype.constructor = Car; // Set the constructor property back to Car

        const car1 = new Car("Volkswagen", "Golf", 2026);
        console.log(car1.getType()); // Output: vehicle
    ```
    - In this example, we have a `Vehicle` constructor with a method `getType`. The `Car` constructor inherits from `Vehicle` by setting its prototype to an object created from `Vehicle.prototype`. This allows instances of `Car` to access the `getType` method defined in `Vehicle`, demonstrating how prototypes can create a chain of inheritance.


3. _That the original object inherits from, has access to all of its `[[Prototype]]` properties and methods:_
    - When an object inherits from a prototype, it has access to all the properties and methods defined on that prototype. This means that if a property or method is not found on the object itself, JavaScript will look up the prototype chain to find it.
    - Example:
    ```javascript
        function Vehicle() {
            this.type = "vehicle";
        }

        Vehicle.prototype.getType = function() {
            return this.type;
        };

        function Car(make, model, year) {
            Vehicle.call(this); // Call the Vehicle constructor to initialize the type property
            this.make = make;
            this.model = model;
            this.year = year;
        }

        Car.prototype = Object.create(Vehicle.prototype); // Set Car's prototype to be an object created from Vehicle's prototype
        Car.prototype.constructor = Car; // Set the constructor property back to Car

        const car1 = new Car("Volkswagen", "Golf", 2026);
        console.log(car1.getType()); // Output: vehicle
    ```
    - In this example, `car1` is an instance of `Car`, which inherits from `Vehicle`. Even though `car1` does not have its own `getType` method, it can access it through the prototype chain, demonstrating that the original object (car1) has access to all of its `[[Prototype]]` properties and methods.

- Why use prototypes?
    1. We can define properties and functions common among all objects on a prototype to save memory. Defining every property and function takes up a lot of memory, especially if you have a lot of common properties and functions, and a lot of created objects! Defining them on a centralized, shared object which the objects have access to, thus saves memory.

    2. The second reason is the name of this section, `Prototypal Inheritance`, which weâ€™ve referred to in passing earlier, in the introduction to the Prototype. In recap, we can say that the `player1` and `player2` objects inherit from the Player.prototype object, which allows them to access functions like `.sayHello`.

### Factory Functions & the Module Pattern

#### Variable Scope

- **Scoping:** the accessibility of variables and functions in different parts of the code. JavaScript has function scope and block scope.
    - **Function scope:** variables declared within a function are only accessible within that function.
    - **Block scope:** variables declared with `let` or `const` within a block (e.g., inside an `if` statement or a loop) are only accessible within that block.
- Example:
    ```javascript
        function exampleFunction() {
            var functionScopedVar = "I am function scoped";
            if (true) {
                let blockScopedVar = "I am block scoped";
                console.log(functionScopedVar); // Output: I am function scoped
                console.log(blockScopedVar); // Output: I am block scoped
            }
            console.log(functionScopedVar); // Output: I am function scoped
            console.log(blockScopedVar); // Error: blockScopedVar is not defined
        }
        exampleFunction();
    ```

- There are 2 types of scope:
    1. **Global Scope:** variables declared outside of any function or block are in the global scope and can be accessed from anywhere in the code.
        - Example: 
        ```javascript
            var globalVar = "I am global";
            function exampleFunction() {
                console.log(globalVar); // Output: I am global
            }
            exampleFunction();
        ```

    2. **Local Scope:** variables declared within a function or block are in the local scope and can only be accessed within that function or block.
        - Example:
        ```javascript
            function exampleFunction() {
                var localVar = "I am local";
                console.log(localVar); // Output: I am local
            }
            exampleFunction();
            console.log(localVar); // Error: localVar is not defined
        ```

#### Closures

- **Closures** are a fundamental concept in JavaScript that allows a function to access variables from its outer (enclosing) scope even after the outer function has finished executing. A closure is created when a function is defined inside another function and retains access to the outer function's variables.
- Example:
    ```javascript
        function outerFunction() {
            let outerVariable = "I am from the outer function";
            
            function innerFunction() {
                console.log(outerVariable); // Output: I am from the outer function
            }
            
            return innerFunction;
        }

        const closureExample = outerFunction();
        closureExample(); // Output: I am from the outer function
    ```
    - Explanation:
        - In this example, `outerFunction` defines a variable `outerVariable` and an inner function `innerFunction` that accesses `outerVariable`.
        - When `outerFunction` is called, it returns `innerFunction`, which retains access to `outerVariable` even after `outerFunction` has finished executing. This is the essence of a closure: the inner function "closes over" the variables of the outer function, allowing it to access and manipulate those variables even when the outer function is no longer in scope.

- **Surrounding State / Lexical Environment:** the variables and functions that are in scope at the time a closure is created. The surrounding state is what allows closures to access and manipulate variables from their outer scope.
- Example:
    ```javascript
        function createCounter() {
            let count = 0; // This is the surrounding state
            
            return function() { // This inner function is the closure
                count++; // Accessing and modifying the surrounding state
                console.log(count);
            };
        }

        const counter = createCounter();
        counter(); // Output: 1
        counter(); // Output: 2
        counter(); // Output: 3
    ```
    - Explanation:
        - In this example, `createCounter` initializes a variable `count` and returns an inner function that increments and logs `count`.
        - The inner function is a closure that has access to the surrounding state (the `count` variable) even after `createCounter` has finished executing. Each time we call `counter()`, it updates and logs the current value of `count`, demonstrating how closures can maintain and manipulate state across multiple invocations.

#### Factory Functions

- **Factory Functions** are functions that create and return new objects. They are a way to create multiple instances of an object with the same structure but different values without using classes or constructors.
- Example:
    ```javascript
        function createCar(make, model, year) {
            return {
                make: make,
                model: model,
                year: year,
                getCarInfo() {
                    return `${this.year} ${this.make} ${this.model}`;
                }
            };
        }

        const car1 = createCar("Volkswagen", "Golf", 2026);
        const car2 = createCar("Toyota", "Camry", 2025);

        console.log(car1.getCarInfo()); // Output: 2026 Volkswagen Golf
        console.log(car2.getCarInfo()); // Output: 2025 Toyota Camry
    ```
    - Explanation:
        - In this example, `createCar` is a factory function that takes parameters to create and return a new car object with properties and a method.
        - Each time we call `createCar`, it returns a new object with the specified properties and the `getCarInfo` method. This allows us to create multiple car objects with different values while sharing the same structure defined by the factory function.

- Advantages of using factory functions over constructors:
    1. Factory functions can return any type of object, not just instances of a specific constructor. This allows for more flexibility in the types of objects that can be created.

    2. Factory functions can have private variables and methods, which are not accessible from outside the function. This allows for better encapsulation and data hiding.
    
    3. Factory functions do not require the use of the `new` keyword, which can help prevent errors and improve readability.

#### Private Variables & Functions

- **Private Variables & Functions** are variables and functions that are not accessible from outside the function or object they are defined in. They are used to encapsulate data and behavior, providing a way to hide implementation details and protect the internal state of an object.

- Example:
    ```javascript
        function createCounter() {
            let count = 0; // This is a private variable
            
            function increment() { // This is a private function
                count++;
                console.log(count);
            }
            
            return {
                increment // Exposing the increment function to the outside
            };
        }

        const counter = createCounter();
        counter.increment(); // Output: 1
        counter.increment(); // Output: 2
        console.log(counter.count); // Output: undefined (count is private)
    ```
    - Explanation:
        - In this example, `count` is a private variable that cannot be accessed directly from outside the `createCounter` function. The `increment` function is also private and can only be called through the object returned by `createCounter`.
        - By using closures, we can create private variables and functions that are encapsulated within the factory function, allowing us to control access to them and protect the internal state of our objects.

#### IIFEs & the Module Pattern

- **Module Pattern** is a design pattern used to create modules in JavaScript. It allows us to encapsulate code and create private variables and functions while exposing only the necessary parts of the module to the outside world. This is often achieved using an Immediately Invoked Function Expression (IIFE).

- **IIEFs (Immediately Invoked Function Expressions)** are functions that are defined and immediately executed. They are often used to create a new scope and avoid polluting the global namespace.
- Example:
    ```javascript
        (function() {
            const privateVariable = "I am private";
            console.log(privateVariable);
        })();
    ```
    - Explanation:
        - In this example, we define an anonymous function and immediately invoke it. This creates a new scope for the `privateVariable`, which is not accessible from outside the function. This pattern is useful for encapsulating code and preventing variables from leaking into the global scope.

### Classes

#### Intro

- JS does not have traditional classes like other programming languages, but it has a class syntax that provides a more familiar way to create objects and handle inheritance. The class syntax is syntactic sugar over the existing prototype-based inheritance in JavaScript, making it easier to write and understand object-oriented code.
- Example:
    ```javascript
        class Car {
            constructor(make, model, year) {
                this.make = make;
                this.model = model;
                this.year = year;
            }

            getCarInfo() {
                return `${this.year} ${this.make} ${this.model}`;
            }
        }

        const car1 = new Car("Volkswagen", "Golf", 2026);
        console.log(car1.getCarInfo()); // Output: 2026 Volkswagen Golf
    ```
    - Explanation:
        - In this example, we define a class `Car` with a constructor that initializes the properties of the car object and a method `getCarInfo` that returns a string with the car's information.
        - We create an instance of the `Car` class using the `new` keyword, which allows us to call the constructor and initialize the object. The class syntax provides a clearer and more concise way to define objects and their behavior compared to using constructor functions and prototypes directly.

- **_Basic class syntax:_**
    ```javascript
        class ClassName {
            constructor(parameters) {
                // initialization code
            }

            methodName() {
                // method code
            }
        }
    ```
- The `constructor` method is a special method for creating and initializing an object created with a class. It is called automatically when a new instance of the class is created using the `new` keyword. The `constructor` can take parameters to initialize the properties of the object.

- _Differences between an object constructor and a class:_
    1. Syntax: Classes use a more concise and familiar syntax compared to constructor functions, making it easier to read and write object-oriented code.

    2. Inheritance: Classes provide a built-in way to handle inheritance using the `extends` keyword, while constructor functions require manual setup of the prototype chain for inheritance.

    3. Static Methods: Classes allow you to define static methods that are called on the class itself, rather than on instances of the class. Constructor functions do not have this feature.

    4. Hoisting: Class declarations are not hoisted, meaning you cannot use a class before it is declared in the code. Constructor functions, on the other hand, are hoisted and can be used before their declaration.

- **Getters and Setters** are special methods in classes that allow you to define how to access and modify the properties of an object. 
    1. Getters are used to retrieve the value of a property, while 
    
    2. setters are used to set the value of a property.

- Example:
    ```javascript
        class Car {
            constructor(make, model, year) {
                this.make = make;
                this.model = model;
                this.year = year;
            }

            get carInfo() { // Getter
                return `${this.year} ${this.make} ${this.model}`;
            }

            set updateYear(newYear) { // Setter
                this.year = newYear;
            }
        }

        const car1 = new Car("Volkswagen", "Golf", 2026);
        console.log(car1.carInfo); // Output: 2026 Volkswagen Golf

        car1.updateYear = 2027; // Using the setter to update the year
        console.log(car1.carInfo); // Output: 2027 Volkswagen Golf
    ```
    - Explanation:
        - In this example, we define a getter `carInfo` that returns a string with the car's information and a setter `updateYear` that allows us to update the year of the car.
        - We can access the getter as if it were a property (`car1.carInfo`) and use the setter to update the year by assigning a new value to it (`car1.updateYear = 2027`). This provides a clean and intuitive way to manage object properties while encapsulating the logic for getting and setting values.

#### Inheritance with Classes

- **_Inheritance_** is a fundamental concept in object-oriented programming that allows a new class (called a `subclass` or `child class`) to inherit properties and methods from an existing class (called a `superclass` or `parent class`). In JavaScript, we can achieve inheritance using the `extends` keyword in class syntax.

- Example:
    ```javascript
        class Vehicle {
            constructor(type) {
                this.type = type;
            }

            getType() {
                return this.type;
            }
        }

        class Car extends Vehicle {
            constructor(make, model, year) {
                super("car"); // Call the parent class constructor
                this.make = make;
                this.model = model;
                this.year = year;
            }
        }
        const car1 = new Car("Volkswagen", "Golf", 2026);
        console.log(car1.getType()); // Output: car (inherited from Vehicle)
    ```
    - Explanation:
        - In this example, we have a `Vehicle` class with a method `getType`. The `Car` class extends `Vehicle`, which means it inherits the properties and methods of `Vehicle`.
        - The `super("car")` call in the `Car` constructor is used to call the constructor of the parent class (`Vehicle`) and initialize the `type` property.
        - When we create an instance of `Car` and call `car1.getType()`, it successfully accesses the inherited method from the `Vehicle` class, demonstrating how inheritance works with classes in JavaScript.

#### Private Class Fields & Methods

- A private class field is a property that is only accessible within the class it is defined in. It is denoted by a `#` prefix before the field name. Private fields cannot be accessed or modified from outside the class, providing encapsulation and data hiding.

- Example:
    ```javascript
        class Car {
            #make; // Private field
            #model; // Private field

            constructor(make, model) {
                this.#make = make;
                this.#model = model;
            }

            getCarInfo() {
                return `${this.#make} ${this.#model}`;
            }
        }

        const car1 = new Car("Volkswagen", "Golf");
        console.log(car1.getCarInfo()); // Output: Volkswagen Golf
        console.log(car1.#make); // Error: Private field '#make' must be declared in an enclosing class
    ```
    - Explanation:
        - In this example, `#make` and `#model` are private fields that cannot be accessed from outside the `Car` class. The `getCarInfo` method can access these private fields because it is defined within the class.
        - Attempting to access `car1.#make` from outside the class results in an error, demonstrating that the private fields are truly private and cannot be accessed or modified from outside the class. This feature allows for better encapsulation and data hiding in JavaScript classes.

- Private methods can also be defined in a similar way using the `#` prefix. They are only accessible within the class and cannot be called from outside the class.

- Example:
    ```javascript
        class Car {
            #make; // Private field
            #model; // Private field

            constructor(make, model) {
                this.#make = make;
                this.#model = model;
            }

            #getCarInfo() { // Private method
                return `${this.#make} ${this.#model}`;
            }

            publicGetCarInfo() { // Public method that calls the private method
                return this.#getCarInfo();
            }
        }

        const car1 = new Car("Volkswagen", "Golf");
        console.log(car1.publicGetCarInfo()); // Output: Volkswagen Golf
        console.log(car1.#getCarInfo()); // Error: Private method '#getCarInfo' must be declared in an enclosing class
    ```
    - Explanation:
        - In this example, `#getCarInfo` is a private method that can only be called from within the `Car` class. The `publicGetCarInfo` method is a public method that calls the private method to retrieve the car information.
        - Attempting to call `car1.#getCarInfo()` from outside the class results in an error, demonstrating that private methods are also encapsulated and cannot be accessed from outside the class. This allows for better control over the internal behavior of the class while still providing a public interface for interacting with it.

#### Static Properties & Methods

- Static properties and methods are associated with the class itself rather than instances of the class. They can be accessed directly on the class without needing to create an instance. Static members are defined using the `static` keyword.

- Example:
    ```javascript
        class Car {
            static numberOfWheels = 4; // Static property

            constructor(make, model) {
                this.make = make;
                this.model = model;
            }

            static getNumberOfWheels() { // Static method
                return Car.numberOfWheels;
            }
        }

        console.log(Car.numberOfWheels); // Output: 4 (accessing static property)
        console.log(Car.getNumberOfWheels()); // Output: 4 (calling static method)
    ```
    - Explanation:
        - In this example, `numberOfWheels` is a static property that belongs to the `Car` class, and `getNumberOfWheels` is a static method that returns the value of the static property.
        - We can access the static property and call the static method directly on the `Car` class without creating an instance. This is useful for properties and methods that are relevant to the class as a whole rather than individual instances.

### Modules

- **Modules:** are a way to organize and encapsulate code in JavaScript. They allow you to break your code into smaller, reusable pieces, and control the visibility of variables and functions. Modules can be implemented using various patterns, such as the Module Pattern or ES6 modules. In simpler terms, modules help you keep your code organized and prevent naming conflicts by encapsulating functionality and exposing only what is necessary to the outside world.

#### Import & Export

- In Modules, we can use `export` to make variables, functions, or classes available for use in other files, and `import` to bring those exported members into another file.
- Example:
    ```javascript
        // In mathUtils.js
        export function add(a, b) {
            return a + b;
        }

        export const PI = 3.14159;

        // In main.js
        import { add, PI } from './mathUtils.js';

        console.log(add(2, 3)); // Output: 5
        console.log(PI); // Output: 3.14159
    ```
    - Explanation:
        - In this example, we have a module `mathUtils.js` that exports a function `add` and a constant `PI`. In another file `main.js`, we import these members using the `import` statement.
        - This allows us to keep our code organized by separating functionality into different files while still being able to use the exported members across our application.

- There are 2 types of importing and exporting:
    1. **Named Exports/Imports:** You can export multiple variables, functions, or classes from a module using named exports. When importing, you need to use the same names as the exported members. There are 2 ways to use `named` exports:
        * Exporting/Importing as you declare:

        ```javascript
            export function add(a, b) {
                return a + b;
            }

            export const PI = 3.14159;
        ```

        ```javascript
        //Importing
            import { add, PI } from './mathUtils.js';
        ```


        * Exporting/Importing after you declare:
        ```javascript
            function add(a, b) {
                return a + b;
            }

            const PI = 3.14159;

            export { add, PI };
        ```

        ```javascript
        //importing
            import { add, PI } from './mathUtils.js';
        ```



    

2. **Default Exports/Imports:** A module can have **_only one_** default export, which can be a variable, function, or class. When importing a default export, you can choose any name for it.
    - Example:
        ```javascript
            // In mathUtils.js
            export default function subtract(a, b) {
                return a - b;
            }

            // In main.js
            import subtract from './mathUtils.js';

            console.log(subtract(5, 2)); // Output: 3
        ```
        - Explanation:
            - In this example, we have a default export of the `subtract` function in `mathUtils.js`. When we import it in `main.js`, we can choose any name for the imported function (in this case, we also called it `subtract` for clarity).
            - Default exports are useful when you want to export a single main functionality from a module, while named exports are better for exporting multiple related functionalities.

    - Another example of default export:
        ```javascript
            // In car.js
            export default class Car {
                constructor(make, model) {
                    this.make = make;
                    this.model = model;
                }
            }

            // In main.js
            import Car from './car.js';

            const myCar = new Car("Volkswagen", "Golf");
            console.log(myCar); // Output: Car { make: 'Volkswagen', model: 'Golf' }
        ```
        - Explanation:
            - In this example, we have a default export of the `Car` class in `car.js`. When we import it in `main.js`, we can create an instance of the `Car` class and use it as needed. This demonstrates how default exports can be used to export classes or other main functionalities from a module.

- You can combine named and default exports in the same module, but you need to be careful with the syntax when importing them.
    - Example:
        ```javascript
            // In mathUtils.js
            export function add(a, b) {
                return a + b;
            }

            export const PI = 3.14159;

            export default function subtract(a, b) {
                return a - b;
            }

            // In main.js
            import subtract, { add, PI } from './mathUtils.js';

            console.log(add(2, 3)); // Output: 5
            console.log(PI); // Output: 3.14159
            console.log(subtract(5, 2)); // Output: 3
        ```
        - Explanation:
            - In this example, we have both named exports (`add` and `PI`) and a default export (`subtract`) in the `mathUtils.js` module. When importing, we first import the default export (which can be named anything) and then the named exports using their specific names. This allows us to use all the exported members from the module in our `main.js` file.

#### Entry Points

- When we use ESM (ECMAScript Modules) instead of adding every JavaScript file to our HTML in order, we only need to link a single file - the **_entry point_**. 
- The `entry point` is the main JavaScript file that serves as the starting point for our application. It is responsible for importing other modules and initializing the application. By using an entry point, we can manage our code more efficiently and keep our HTML cleaner, as we only need to include one script tag that points to the entry point file.

- We mostly use the `index.js` file as the entry point, but it can be named anything. The important thing is that it serves as the main file that imports other modules (dependent on other files) and starts the application.

- Declaration: 
    ```html
        <script type="module" src="index.js"></script>
    ```
    - Explanation:
        - In this example, we have an `index.js` file that serves as the entry point for our application. We declare it in our HTML using a `<script>` tag with the `type="module"` attribute, which tells the browser that this script is an ES6 module.
        - The `index.js` file can then import other modules and initialize the application. This approach allows us to keep our HTML clean and organized, as we only need to include one script tag for the entry point, and it also helps us manage our code more efficiently by breaking it into smaller, reusable modules.

- **_Dependency Graph:_** When we have multiple modules that depend on each other, we can visualize the relationships between them using a dependency graph. 

- A dependency graph is a directed graph that shows how modules are connected based on their import and export relationships. Each node in the graph represents a module, and each directed edge represents an import statement from one module to another.

- Example: 
    ```
        index.js
        â”śâ”€â”€ mathUtils.js
        â”‚   â”śâ”€â”€ add()
        â”‚   â””â”€â”€ PI
        â””â”€â”€ car.js
            â””â”€â”€ Car class
    ```
    - Explanation:
        - In this example, we have an `index.js` file that serves as the entry point for our application. It imports two modules: `mathUtils.js` and `car.js`.
        - The `mathUtils.js` module exports a function `add()` and a constant `PI`, while the `car.js` module exports a `Car` class. The dependency graph shows how these modules are connected, with `index.js` depending on both `mathUtils.js` and `car.js`. This visualization helps us understand the structure of our code and how different modules interact with each other.

#### Common JS (CJS)

- CJS (CommonJS) is a module system used in Node.js for organizing and structuring code. It allows developers to create reusable modules and manage dependencies in a modular way. In CommonJS, we use `require` to import modules and `module.exports` to export them.
- Example:
    ```javascript
        // In mathUtils.js
        function add(a, b) {
            return a + b;
        }

        const PI = 3.14159;

        module.exports = { add, PI };

        // In main.js
        const { add, PI } = require('./mathUtils.js');

        console.log(add(2, 3)); // Output: 5
        console.log(PI); // Output: 3.14159
    ```
    - Explanation:
        - In this example, we define a module `mathUtils.js` that exports a function `add` and a constant `PI` using `module.exports`. In another file `main.js`, we import these members using the `require` function.
        - CommonJS is primarily used in Node.js environments and is not natively supported in browsers. However, tools like Browserify or Webpack can be used to bundle CommonJS modules for use in the browser.

### npm

- **npm:** is a package manager for JavaScript that allows developers to easily share and reuse code. npm provides a vast ecosystem of open-source libraries and tools that can be easily installed and managed in your projects. With npm, you can install packages from the npm registry, manage dependencies, and run scripts to automate tasks in your development workflow.

- In simpler terms, `npm` is a tool that helps you find and use code written by other developers, making it easier to build your applications without having to reinvent the wheel.

- To use npm, you need to have `Node.js` installed on your machine, as npm comes bundled with Node.js. Once you have Node.js installed, you can use npm to manage your project dependencies and run scripts.

#### package.json

- `package.json` is a file that contains metadata about your project and its dependencies. It is used by npm to manage your project's dependencies, scripts, and other configurations. The `package.json` file is typically located in the root directory of your project. It includes information such as the project name, version, description, author, license, and most importantly, the dependencies and scripts.

- Itâ€™s a JSON _(JavaScript Object Notation)_ file containing information about our project, such as its name or any dependencies and their version numbers. npm can read this file and do things such as install all of the listed dependencies with the correct versions, and running commands that youâ€™ve set as an npm script

- Example: 
    ```json
        {
            "name": "my-project",
            "version": "1.0.0",
            "description": "A sample project using npm",
            "main": "index.js",
            "scripts": {
                "start": "node index.js",
                "test": "echo \"Error: no test specified\" && exit 1"
            },
            "author": "Your Name",
            "license": "ISC",
            "dependencies": {
                "express": "^4.17.1"
            }
        }
    ```
    - Explanation:
        - In this example, we have a `package.json` file that defines a project with the name `my-project`, version `1.0.0`, and a description. It specifies `index.js` as the main entry point of the application.
        
        - The `scripts` section defines two scripts: `start`, which runs the application using Node.js, and `test`, which is a placeholder for running tests.

        - The `dependencies` section lists the packages that the project depends on, in this case, the `express` package with a version of `^4.17.1`. This means that when we run `npm install`, npm will install the specified version of Express and its dependencies.

        - The `license` field specifies the license under which the project is released, and the `author` field provides information about the creator of the project. Overall, the `package.json` file serves as a central configuration file for managing the project's dependencies and scripts.

- Read more here: [the package.json file](https://docs.npmjs.com/creating-a-package-json-file)

##### Installing packages with npm

- npm is used to install packages _(dependencies, or in simpler terms, libraries)_ from the npm registry, which is a large collection of open-source libraries and tools. You can install packages globally or locally in your project.

- To install a package locally in your project, you can use the following command:
    ```bash
        npm install package-name
    ```
    - This will install the specified package and add it to the `dependencies` section of your `package.json` file.

- Read more here: [installing packages with npm](https://docs.npmjs.com/downloading-and-installing-packages-locally) 

### Webpack

1. [Webpack concepts](https://webpack.js.org/concepts/)

2. [Webpack Asset Management Guide](https://webpack.js.org/guides/asset-management/)

#### Bundling

- **Bundling:** is the process of taking multiple JavaScript files and combining them into a single file (or a few files) that can be included in an HTML page. This is important for optimizing the loading of web applications, as it reduces the number of HTTP requests needed to load all the necessary JavaScript files. 

- Bundling also allows you to use modern JavaScript features and modules in your code, as the bundler will take care of resolving dependencies and ensuring that the code runs correctly in the browser.

- We provide an `entry point` (the main JavaScript file that serves as the starting point for our application) to the bundler, and it will analyze the dependencies of that file and include all the necessary modules in the final bundle.

- Example:
    ```javascript
        // index.js (entry point)
        import { add } from './mathUtils.js';
        console.log(add(2, 3)); // Output: 5

        // mathUtils.js
        export function add(a, b) {
            return a + b;
        }
    ```
    - Explanation:
        - In this example, we have an `index.js` file that serves as the entry point for our application. It imports the `add` function from `mathUtils.js`.
        
        - When we run the `bundler` (e.g., **Webpack**), it will analyze the dependencies of `index.js`, find that it depends on `mathUtils.js`, and include both files in the final bundle. This allows us to write modular code using ES6 modules while still being able to run it in the browser without worrying about module support or multiple script tags.

#### Webpack

- **Webpack:** is a popular module bundler for JavaScript applications. It takes your JavaScript files and their dependencies, processes them, and generates a single bundle (or multiple bundles) that can be included in your HTML page. Webpack also provides features like code splitting, hot module replacement, and support for various loaders and plugins to enhance the development experience.

##### Bundling JS using Webpack

1. **_Step 1: Install Webpack and Webpack CLI as development dependencies in your project:_**
    ```bash
        npm install --save-dev webpack webpack-cli
    ```

    - This command installs Webpack and its command-line interface (CLI) as development dependencies, which means they are only needed during the development process and will not be included in the production build.

    - It also creates a `node_modules` directory in your project where the Webpack packages are installed, and it adds an entry for Webpack in the `devDependencies` section of your `package.json` file.

    - It also creates a `package-lock.json` file if it doesn't already exist, which locks the versions of the installed packages to ensure consistent installations across different environments.

* **src & dist:**
    - `src` (source) is the directory where you typically place your original JavaScript files and other assets that you want to bundle. This is where you write your code and organize it in a modular way.

    - `dist` (distribution) is the directory where the bundled output from Webpack is generated. This is the directory that you would include in your HTML page to load the bundled JavaScript file(s). The `dist` directory is usually specified in the Webpack configuration file, and it is where Webpack will output the final bundled files after processing your source code from the `src` directory.

    - In simpler terms, `src` is where you write your code, and `dist` is where the bundled version of your code is generated for use in production. That is, you `deploy` the contents of the `dist` directory to your web server to make your application available to users.

    - If you clone or fork a project that uses Webpack, yyou only need to run `npm install` to install the dependencies, and then you can run the Webpack build process to generate the bundled files in the `dist` directory. You do not need to manually create the `dist` directory, as Webpack will handle that for you when it generates the output.

2. **_Step 2: Create a `src` directory and add your JavaScript files there._** 
- For example, create an `src` with 2 js files inside: `index.js` and `greeting.js`:

    - `src/index.js`:
        ```javascript
            import { greeting } from "./greeting.js";
            console.log(greeting);

        ```

    - `src/greeting.js`:
        ```javascript
            export const greeting = "Hello, Odinite!";
        ```
    - Explanation:
        - In this step, we create a `src` directory where we place our original JavaScript files. We have an `index.js` file that serves as the entry point for our application, and it imports (depends on) a `greeting` constant from `greeting.js`. The `greeting.js` file exports a constant string that we want to use in our application.

3. **_Step 3: Create a `webpack.config.js` file in the root directory of your project to configure Webpack:_**
    ```javascript
        // webpack.config.js
            import path from "node:path"; // Importing the path module to resolve file paths

            export default {
                mode: "development", // Setting the mode to development for better debugging and readable output

                entry: "./src/index.js", // Specifying the entry point for the application

                output: {
                    filename: "main.js", // The name of the output bundle file

                    path: path.resolve(import.meta.dirname, "dist"), // Resolving the output path to the 'dist' directory

                    clean: true, // Clean the output directory before each build to remove old files
                },
            };
    ```
    - Explanation:
        - In this step, we create a `webpack.config.js` file to configure Webpack. We set the `mode` to `development` for better debugging and readable output during development.
        
        - We specify the `entry` point of our application as `./src/index.js`, which is the main JavaScript file that serves as the starting point for our application.

        - In the `output` section, we define the name of the output bundle file as `main.js` and resolve the output path to the `dist` directory using the `path` module. The `clean: true` option ensures that the output directory is cleaned before each build, removing any old files that may be present.


4. **_Step 4: Run the Webpack build process to generate the bundled files in the `dist` directory:_**

    ```bash
        npx webpack
    ```
    - Explanation:
        - In this step, we run the Webpack build process using the command `npx webpack`. This command tells Webpack to read the configuration from `webpack.config.js`, process the files starting from the specified entry point (`./src/index.js`), and generate the bundled output in the `dist` directory as defined in the configuration.

        - After running this command, you should see a `main.js` file in the `dist` directory, which is the bundled version of your application that can be included in your HTML page for production use.

- If you run `node dist/main.js`, you will see the output of the bundled code in the terminal, which should log "Hello, Odinite!" to the console, demonstrating that the bundling process has successfully combined the code from `index.js` and `greeting.js` into a single file.

5. **_Step 5: Include the bundled file in your HTML page:_**

    ```html
        <!DOCTYPE html>
        <html lang="en">
        <head>
            <meta charset="UTF-8">
            <meta name="viewport" content="width=device-width, initial-scale=1.0">
            <title>Webpack Example</title>
        </head>
        <body>
            <script src="./dist/main.js"></script> <!-- Including the bundled JavaScript file -->
        </body>
        </html>
    ```
    - Explanation:
        - In this final step, we include the bundled JavaScript file (`main.js`) in our HTML page using a `<script>` tag. This allows us to load the bundled code into the browser and run our application.

        - The `src` attribute of the `<script>` tag points to the location of the bundled file in the `dist` directory. When the browser loads this HTML page, it will execute the code in `main.js`, which contains all the necessary JavaScript for our application as processed by Webpack.

##### Bundling HTML using Webpack

1. **_Step 1: Install Webpack and Webpack CLI as development dependencies in your project:_**

    ```bash
        npm install --save-dev webpack webpack-cli
    ```

2. **_Step 2: Install the `html-webpack-plugin` to handle HTML files in Webpack:_**

    ```bash
        npm install --save-dev html-webpack-plugin
    ```

    - This command installs the `html-webpack-plugin`, which is a plugin for Webpack that simplifies the creation of HTML files to serve your bundled JavaScript. **It automatically injects the bundled JavaScript files into the HTML file, making it easier to manage your HTML and JavaScript together. Therefore, no need to manually include the bundled file in your HTML.**

    - It also adds an entry for `html-webpack-plugin` in the `devDependencies` section of your `package.json` file.

    - It also creates a `package-lock.json` file if it doesn't already exist, which locks the versions of the installed packages to ensure consistent installations across different environments.

3. **_Step 3: Create a `template.html` file in the `src` directory that will serve as the template for your HTML file:_**
- (you can name this file anything you like)

    ```html
        <!-- src/template.html -->
        <!DOCTYPE html>
        <html lang="en">
        <head>
            <meta charset="UTF-8">
            <meta name="viewport" content="width=device-width, initial-scale=1.0">
            <title>Webpack HTML Example</title>
        </head>
        <body>
            <h1>Welcome to Webpack!</h1>
        </body>
        </html>
    ```
    - Explanation:
        - In this step, we create a `template.html` file in the `src` directory that serves as the template for our HTML file. This template will be used by the `html-webpack-plugin` to generate the final HTML file that includes the bundled JavaScript.

4. **_Step 4: Create a `webpack.config.js` file in the root directory of your project to configure Webpack and use the `html-webpack-plugin`:_**

    ```javascript
        // webpack.config.js
        import path from "node:path";
        import HtmlWebpackPlugin from "html-webpack-plugin";

        export default {
            mode: "development",
            entry: "./src/index.js",
            output: {
                filename: "main.js",
                path: path.resolve(import.meta.dirname, "dist"),
                clean: true,
            },
            
            plugins: [
                new HtmlWebpackPlugin({
                template: "./src/template.html",
                }),
            ],
        };
    ```
    - Explanation:
        - In this step, we configure Webpack to use the `html-webpack-plugin`. We import the plugin at the top of the configuration file and then add it to the `plugins` array in the Webpack configuration.

        - The `HtmlWebpackPlugin` is instantiated with an options object that specifies the `template` property, which points to our `template.html` file. This tells the plugin to use our template as the basis for generating the final HTML file.

        - When we run the Webpack build process, the `html-webpack-plugin` will take our `template.html`, inject the bundled JavaScript file (`main.js`) into it, and output a new HTML file in the `dist` directory that can be served to users. This allows us to manage our HTML and JavaScript together without having to manually include script tags in our HTML.


- If we run `npx webpack`, Webpack will generate a new HTML file in the `dist` directory that includes the bundled JavaScript file. The generated HTML file will be based on our `template.html` and will have a script tag automatically injected to include the bundled JavaScript. This allows us to easily manage our HTML and JavaScript together without having to manually update our HTML file every time we change our JavaScript code.

##### Loading CSS with Webpack

- For CSS, we need 2 packages instead of 1:
    1. `style-loader`: Injects CSS into the DOM by adding a `<style>` tag. In simpler terms, it takes the CSS code and adds it to the HTML document so that the styles are applied to the page.

    2. `css-loader`: Interprets `@import` and `url()` like `import/require()` and resolves them. In simpler terms, it allows you to import CSS files into your JavaScript files and handles the dependencies between them.

- The `css-loader` will read any CSS files we import in a JS file and store the result in a string. The `style-loader` then takes that string and actually adds the JS code that will apply those styles to the page. Thus the need for both.

1. **_Step 1: Install the necessary loaders for handling CSS files in Webpack:_**

    ```bash
        npm install --save-dev style-loader css-loader
    ```

    - This command installs both `style-loader` and `css-loader` as development dependencies in your project. These loaders are essential for processing CSS files when bundling with Webpack.

    - It also adds entries for both loaders in the `devDependencies` section of your `package.json` file.

    - It also creates a `package-lock.json` file if it doesn't already exist, which locks the versions of the installed packages to ensure consistent installations across different environments.

2. **_Step 2: Add these loaders to your Webpack configuration to handle CSS files:_**
- Since these aren't plugins but loaders, they go in a separate section:

    ```javascript
        // webpack.config.js
        import path from "node:path";
        import HtmlWebpackPlugin from "html-webpack-plugin";

        export default {
            mode: "development",
            entry: "./src/index.js",
            output: {
                filename: "main.js",
                path: path.resolve(import.meta.dirname, "dist"),
                clean: true,
            },
            plugins: [
                new HtmlWebpackPlugin({
                template: "./src/template.html",
                }),
            ],
            module: {
                rules: [
                {
                    test: /\.css$/i,
                    use: ["style-loader", "css-loader"],
                },
                ],
            },
        };
    ```

    - Explanation:
        - In this step, we update our Webpack configuration to include rules for handling CSS files. We add a `module` section with a `rules` array that specifies how to process files that match the `.css` extension.

        - The rule uses a regular expression (`test: /\.css$/i`) to identify CSS files and specifies that they should be processed using both `style-loader` and `css-loader`. The order of the loaders is important: `css-loader` processes the CSS file first, and then `style-loader` injects the styles into the DOM.

        - With this configuration, when we import a CSS file in our JavaScript code, Webpack will use these loaders to handle the CSS and ensure that the styles are applied correctly in the browser.

- **_NB: Loader order matters for CSS!_** 
    * _Notice how we put `css-loader` at the end of the array. We must set this order and not the reverse._

    * _Webpack will run the loaders starting at the end, so we want it to read the CSS file into a string with `css-loader` **first**, then use `style-loader` to inject the JavaScript that applies the CSS in that string to the page. It wouldnâ€™t work the same the other way round._


3. **_Step 3: Import the CSS file in your JavaScript entry point (e.g., `index.js`):_**

    ```javascript
        // src/index.js
        import "./styles.css"; // Importing the CSS file

        console.log("Hello, Webpack with CSS!");
    ```

    - Explanation:
        - In this step, we import the CSS file (`styles.css`) into our JavaScript entry point (`index.js`). This tells Webpack to process the CSS file using the loaders we configured in the previous step.

        - When we run the Webpack build process, it will use `css-loader` to read the contents of `styles.css` and then use `style-loader` to inject those styles into the DOM. This allows us to manage our styles directly from our JavaScript files and ensures that they are included in the final bundled output.

4. **_Step 4: Bundle everything with webpack:_**

    ```bash
        npx webpack
    ```

    - Explanation:
        - In this final step, we run the Webpack build process using the command `npx webpack`. This will process all our JavaScript and CSS files according to the configuration we set up, and generate the bundled output in the `dist` directory.

        - After running this command, you should see a new HTML file in the `dist` directory that includes the bundled JavaScript and has the styles from `styles.css` applied to it. This demonstrates that Webpack has successfully processed both our JavaScript and CSS files and generated a complete bundle for our application.

##### What about link tags?

- Should we still link our CSS files in the HTML with `<link>` tags when using Webpack to bundle our CSS?
    - **_No, when using Webpack to bundle your CSS, you do not need to link your CSS files in the HTML with `<link>` tags. Instead, you should import your CSS files directly into your JavaScript entry point (e.g., `index.js`) using an import statement like `import "./styles.css";`._**

##### Loading Images

- Local image files also require extra configuration since they are not JS files.

- There are 3 different ways to deal with local image files:

    1. **_Image files used in our CSS inside `url()`:_**
        * The `css-loader` will automatically handle image files that are referenced in our CSS files using `url()`. When we import a CSS file that contains `url()` references to images, the `css-loader` will process those references and include the images in the bundle. This means that we can simply reference our images in our CSS files, and Webpack will take care of bundling them correctly. Thus, nothing extra to do.

        * Example:
            ```css
                /* styles.css */
                .background {
                    background-image: url('./images/background.jpg');
                }
            ```
            - Explanation:
                - In this example, we have a CSS file that references an image using `url()`. When we import this CSS file in our JavaScript entry point, the `css-loader` will automatically process the image reference and include it in the bundle. This allows us to manage our images directly from our CSS files without needing to do any additional configuration for handling images in Webpack.

    2. **_Image files we reference in out HTML template, e.g. as the src of an `<img>`:_**
        * For images that we reference directly in our HTML template (e.g., using `<img src="./images/logo.png">`), we need to use the `html-loader` to process those image references. The `html-loader` will analyze the HTML file, find any image references, and include those images in the bundle.

        * Without this, a `src` like `./odin.png` would just be a bit of text that will no longer reference the correct file once we run Webpack to build into dist. Letâ€™s install it:

        * To set this up, we need to install the `html-loader` and configure it in our Webpack configuration file.

        * Example:
            ```bash
                npm install --save-dev html-loader
            ```

            - Then, add the following object to the `module.rules` array within `webpack.config.js`:

            ```javascript
                // webpack.config.js
                {
                    test: /\.html$/i,
                    use: ["html-loader"],
                }
            ```
            - Explanation:
                - In this example, we install the `html-loader` and configure it in our Webpack configuration file to handle HTML files. The `html-loader` will process any HTML files that we use as templates (e.g., `template.html`) and will analyze the content for any image references (like `<img src="./images/logo.png">`).

                - When we run the Webpack build process, the `html-loader` will find these image references in our HTML template and include those images in the bundle. This allows us to manage our images directly from our HTML files without needing to do any additional configuration for handling images in Webpack.


    3. **_Images we use in our JS, where we will need to import the files:_**
        - If we need to use a local image file in our JavaScript (such as when manipulating the DOM to create or edit `img` elements and set their `src` attribute), we need to import the images into our JavaScript module. Since images arenâ€™t JavaScript, we need to tell Webpack that these files will be assets by adding an `asset/resource` rule. No need to install anything here. Just add the following object to the `module.rules` array within `webpack.config.js`:

        ```javascript
            // webpack.config.js
            {
                test: /\.(png|svg|jpg|jpeg|gif)$/i,
                type: "asset/resource",
            }
        ```
        - Explanation:
            - In this example, we add a rule to our Webpack configuration to handle image files that we want to use in our JavaScript code. The rule uses a regular expression to match common image file extensions (like `.png`, `.jpg`, etc.) and specifies that these files should be treated as resources using `type: "asset/resource"`.

            - When we import an image file in our JavaScript code (e.g., `import logo from './images/logo.png';`), Webpack will process the image file according to this rule and include it in the bundle. This allows us to manage our images directly from our JavaScript files and ensures that they are included correctly in the final output when we run the Webpack build process.

            - _**PS**: You can always edit the regex in the test property to remove any file extensions you donâ€™t need or add any extensions you do need. _


- Then, in whatever JavaScript module we want to use that image in, we just have to default import it.

```javascript
    // src/index.js
    import odinImage from "./odin.png";

    const image = document.createElement("img");
    image.src = odinImage;

    document.body.appendChild(image);
```

- Explanation:
    - In this example, we import an image file (`odin.png`) into our JavaScript module using a default import. The imported value (`odinImage`) will be a string that represents the path to the image file in the bundled output.

    - We then create a new `img` element using `document.createElement("img")`, set its `src` attribute to the imported image path, and append it to the body of the document. This allows us to dynamically add images to our web page using JavaScript while ensuring that the images are correctly processed and included in the Webpack bundle.

- Our final `webpack.config.js` file should look like this:

```javascript
    // webpack.config.js
    import path from "node:path";
    import HtmlWebpackPlugin from "html-webpack-plugin";

    export default {
        mode: "development",
        entry: "./src/index.js",
        output: {
            filename: "main.js",
            path: path.resolve(import.meta.dirname, "dist"),
            clean: true,
        },
        plugins: [
            new HtmlWebpackPlugin({
            template: "./src/template.html",
            }),
        ],
        module: {
            rules: [
            {
                test: /\.css$/i,
                use: ["style-loader", "css-loader"],
            },
            {
                test: /\.html$/i,
                use: ["html-loader"],
            },
            {
                test: /\.(png|svg|jpg|jpeg|gif)$/i,
                type: "asset/resource",
            },
            ],
        },
    };
```


- You might notice that when images are included when bundling, the output image file in `dist` has a different file name (it will likely be some jumble of numbers and letters). By default, Webpack gives your bundled image files a new name by hashing their contents. You do not need to know how this works, nor do you need to dig into the details of why, nor how to change it. You just need to be aware that this is expected behavior (itâ€™s to do with preventing issues with the browser cache and matching file names).

##### Summary of Webpack:

- **_You only need to configure what you need_**
    * You may not need everything weâ€™ve mentioned. If your project does not have images with local file path sources in your HTML template, you do not need `html-loader` set up. If you arenâ€™t using any local images in your JavaScript, you wonâ€™t need the image `asset/resource` rule set up.

    * Similarly, in the future, you may end up working with things that need a special loader or plugin, such as custom fonts or preprocessors. You can always use Google or reference Webpackâ€™s documentation for instructions on what youâ€™d need when that time comes.

#### Webpack dev server

- The `webpack-dev-server` is a development server that provides live reloading and hot module replacement (HMR) for your Webpack projects. It allows you to serve your bundled application locally and automatically refresh the browser whenever you make changes to your code, providing a faster and more efficient development experience.

- It works the same as the `Live Server` extension in VS Code, but itâ€™s built specifically for Webpack projects and offers additional features like HMR, which allows you to update modules in the browser without a full page reload.

- It ensures you don't run `npx webpack` every time you make a change to your code. Instead, it watches for changes and automatically rebuilds and refreshes the browser for you.

1. **_Step 1: Install the `webpack-dev-server` as a development dependency in your project:_**

    ```bash
        npm install --save-dev webpack-dev-server
    ```

    - This command installs the `webpack-dev-server` package as a development dependency in your project. The `webpack-dev-server` provides a local development server with live reloading and hot module replacement features.

    - It also adds an entry for `webpack-dev-server` in the `devDependencies` section of your `package.json` file.

    - It also creates a `package-lock.json` file if it doesn't already exist, which locks the versions of the installed packages to ensure consistent installations across different environments.

- We can also use `source map`s with the dev server to make debugging easier. Source maps allow you to see the original source code in the browser's developer tools, even though the code has been bundled and minified by Webpack. To enable source maps, you can add the `devtool` property to your Webpack configuration:

```javascript
    // webpack.config.js
    export default {
        // ... other configuration options ...
        devtool: "inline-source-map", // Enable source maps for easier debugging 
        //or
        devtool: "eval-source-map", // Another option for source maps. This option is faster for development but may not provide as detailed source maps as "inline-source-map".
    };
```

2. **_Step 2: Add these properties to the `webpack.config.js` file to configure the development server:_**

    ```javascript
        // webpack.config.js
        import path from "node:path";
        import HtmlWebpackPlugin from "html-webpack-plugin";

        export default {
            mode: "development",
            entry: "./src/index.js",
            output: {
                filename: "main.js",
                path: path.resolve(import.meta.dirname, "dist"),
                clean: true,
            },
            devtool: "eval-source-map",
            devServer: {
                watchFiles: ["./src/template.html"],
            },
            plugins: [
                new HtmlWebpackPlugin({
                template: "./src/template.html",
                }),
            ],
            module: {
                rules: [
                {
                    test: /\.css$/i,
                    use: ["style-loader", "css-loader"],
                },
                {
                    test: /\.html$/i,
                    use: ["html-loader"],
                },
                {
                    test: /\.(png|svg|jpg|jpeg|gif)$/i,
                    type: "asset/resource",
                },
                ],
            },
        };


3. Step 3: Run the development server using the following command:

    ```bash
        npx webpack serve
    ```

    - Explanation:
        - In this step, we run the Webpack development server using the command `npx webpack serve`. This command starts the development server and serves our bundled application locally.

        - The `devServer` configuration in our Webpack config specifies that the server should watch for changes to the `template.html` file. This means that if we make any changes to our HTML template, the development server will automatically rebuild the bundle and refresh the browser to reflect those changes.

        - With the development server running, we can make changes to our source code (JavaScript, CSS, HTML) and see those changes reflected in real-time in the browser without needing to manually refresh or rerun the build process. This provides a much faster and more efficient development experience.

- Our site will then be available at `http://localhost:8080` (or another port if 8080 is already in use). Whenever we make changes to our source files, the development server will automatically rebuild the bundle and refresh the browser to show the latest changes. This allows us to see our updates in real-time without needing to manually refresh the page or rerun the build process.

**_NB: Restart the dev server upon config changes_**
- _Note that the webpack-dev-server only reads your webpack configuration when you start it. If you change the webpack config file while the dev server is running, it will not reflect those config changes. Use Ctrl + C in the terminal to kill it then rerun npx webpack serve to apply the new config._
