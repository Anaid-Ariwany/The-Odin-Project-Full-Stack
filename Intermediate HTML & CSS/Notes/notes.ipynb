### Emmet

- Emmet is a plugin, built into VS Code, that helps you write HTML and CSS more efficiently by providing a bunch of clever shortcuts.

- Find more on https://docs.emmet.io/abbreviations/syntax/

- Example shortcuts:
  - `!` + `Tab` creates a boilerplate HTML document.
  - `ul>li*5` + `Tab` creates an unordered list with five list items.
  - `div#container>div.row>div.col*3` + `Tab` creates a div with an id of "container", which contains a div with a class of "row", which contains three divs with a class of "col".

### SVG

#### Intro

- **_SVG_** stands for **Scalable Vector Graphics**. It is a markup language for describing two-dimensional graphics applications and images, and a set of related graphics script interfaces. SVG is used to define vector-based graphics for the Web, and it can be styled with CSS and scripted with JavaScript.

- In shorter terms: SVGs are a scalable image format, which means they will easily scale to any size and retain their quality without increasing their filesize. They’re also very useful if you need to create or modify your images programmatically, because you can change their properties through CSS and JavaScript.

- SVGs are defined using **_XML (Extensible Markup Language)_**, which means they can be created and edited with any text editor, as well as with drawing software. They can also be animated and interactive, making them a powerful tool for web design and development.

#### Anatomy of an SVG

- Example syntax:
    ```html
    <svg xmlns="http://www.w3.org/2000/svg" viewBox="0 0 100 100">
        <rect x=0 y=0 width=100 height=50 />
        <circle class="svg-circle" cx="50" cy="50" r="10"/>
    </svg>
    ```

1. `xlmns` attribute: This specifies the XML namespace for the SVG document. It is required for the SVG to be properly recognized and rendered by browsers. In simpler words, it tells the browser that this is an SVG document and how to interpret it.

2. `viewBox` attribute: This defines the coordinate system and dimensions of the SVG canvas. It takes four values: `min-x`, `min-y`, `width`, and `height`. In this example, it sets the origin of the coordinate system at (0, 0) and defines a canvas that is 100 units wide and 100 units tall.

3. `<rect>` element: This creates a rectangle shape in the SVG. The `x` and `y` attributes specify the position of the top-left corner of the rectangle, while the `width` and `height` attributes define its dimensions.

4. `<circle>` element: This creates a circle shape in the SVG. The `cx` and `cy` attributes specify the center of the circle, while the `r` attribute defines its radius. The `class` attribute can be used to apply CSS styles to the circle, allowing for further customization of its appearance.

- Find more on: https://developer.mozilla.org/en-US/docs/Web/SVG/Reference/Element 

#### Embedding SVGs

- There are 2 main ways to embed SVGs in your HTML:
    1. **Inline SVG:** You can include the SVG code directly within your HTML document, as shown in the example above. This allows for easy styling and scripting of the SVG elements using CSS and JavaScript. Exmple:
        ```html
            <svg xmlns="http://www.w3.org/2000/svg" viewBox="0 0 100 100">
                <rect x=0 y=0 width=100 height=50 />
                <circle class="svg-circle" cx="50" cy="50" r="10"/>
            </svg>
        ```

    2. **External/Linking SVG:** You can reference an external SVG file using the `<img>` tag or the `<object>` tag. This approach is useful when you want to reuse the same SVG in multiple places or when the SVG is complex and large. Example:
        ```html
            <img src="path/to/your/image.svg" alt="Description of the image">
        ```

### Tables

#### Intro

##### Basics

- Tables are a way to display data in a structured format, using rows and columns. They are created using the `<table>` element in HTML, along with various other elements to define the structure and content of the table.

- To create a table, you typically use the following elements:
    - `<table>`: This is the container element for the entire table.

    - `<th>`: This stands for **"table header"** and is used to define a header cell in the table. It is typically used in the first row of the table to label the columns.

    - `<tr>`: This stands for **"table row"** and is used to define a row in the table.

    - `<td>`: This stands for **"table data"** and is used to define a standard cell in the table that contains data.

- Example of a simple table:
    ```html
    <table>
        <tr>
            <th>Name</th>
            <th>Age</th>
            <th>City</th>
        </tr>
        <tr>
            <td>Alice</td>
            <td>30</td>
            <td>New York</td>
        </tr>
        <tr>
            <td>Bob</td>
            <td>25</td>
            <td>Los Angeles</td>
        </tr>
    </table>
    ```

##### Advanced

- `<thead>`: This element is used to group the header content in a table. It typically contains one or more `<tr>` elements that define the header rows of the table.

- `<tbody>`: This element is used to group the body content in a table. It typically contains one or more `<tr>` elements that define the body rows of the table.

- `<tfoot>`: This element is used to group the footer content in a table. It typically contains one or more `<tr>` elements that define the footer rows of the table.

- Example of a table with `<thead>`, `<tbody>`, and `<tfoot>`:
    ```html
    <table>
        <thead>
            <tr>
                <th>Name</th>
                <th>Age</th>
                <th>City</th>
            </tr>
        </thead>
        <tbody>
            <tr>
                <td>Alice</td>
                <td>30</td>
                <td>New York</td>
            </tr>
            <tr>
                <td>Bob</td>
                <td>25</td>
                <td>Los Angeles</td>
            </tr>
        </tbody>
        <tfoot>
            <tr>
                <td colspan="3">End of Table</td>
            </tr>
        </tfoot>
    </table>
    ```

#### Spanning Cells Across Multiple Rows & Columns

- `colspan` attribute: This is used to make a cell span across multiple columns. You can specify the number of columns you want the cell to span by setting the value of `colspan`. For example, if you want a cell to span across 3 columns, you would use `colspan="3"`. Example:
    ```html
    <table>
        <tr>
            <td colspan="2">Alice</td>
            <td>New York</td>
        </tr>
    </table>
    ```

- `rowspan` attribute: This is used to make a cell span across multiple rows. You can specify the number of rows you want the cell to span by setting the value of `rowspan`. For example, if you want a cell to span across 2 rows, you would use `rowspan="2"`. Example:
    ```html
    <table>
        <tr>
            <td rowspan="2">Alice</td>
            <td>30</td>
            <td>New York</td>
        </tr>
    </table>
    ```

#### Grouping Columns

- `<colgroup>` element: This is used to group columns in a table. It allows you to apply styles or attributes to a group of columns at once. You can specify the number of columns in the group using the `span` attribute. Example:
    ```html
    <table>
        <colgroup>
            <col span="2" style="background-color: lightgray;">
            <col style="background-color: lightblue;">
        </colgroup>
        <tr>
            <th>Name</th>
            <th>Age</th>
            <th>City</th>
        </tr>
        <tr>
            <td>Alice</td>
            <td>30</td>
            <td>New York</td>
        </tr>
    </table>
    ```
    - In this example, the first two columns `(Name and Age)` are grouped together and styled with a light gray background, while the third column `(City)` is styled with a light blue background.



- `<col>` element: This is used within a `<colgroup>` to define individual columns. You can apply styles or attributes to specific columns using the `<col>` element. Example:
    ```html
    <table>
        <colgroup>
            <col style="background-color: lightgray;">
            <col style="background-color: lightblue;">
            <col style="background-color: lightgreen;">
        </colgroup>
        <tr>
            <th>Name</th>
            <th>Age</th>
            <th>City</th>
        </tr>
        <tr>
            <td>Alice</td>
            <td>30</td>
            <td>New York</td>
        </tr>
    </table>
    ```
    - In this example, each column is defined with a different background color using the `<col>` element within the `<colgroup>`.

#### Captioning Tables

- `<caption>` element: This is used to provide a caption or title for the table. It is typically placed immediately after the opening `<table>` tag. The content of the `<caption>` element is displayed above the table by default, but it can be styled with CSS to change its position and appearance. Example:
    ```html
    <table>
        <caption>Employee Information</caption>
        <tr>
            <th>Name</th>
            <th>Age</th>
            <th>City</th>
        </tr>
        <tr>
            <td>Alice</td>
            <td>30</td>
            <td>New York</td>
        </tr>
    </table>
    ```
    - In this example, the caption "Employee Information" is displayed above the table, providing context for the data contained within the table.

#### Scope Attribute

- The `scope` attribute is used in table header cells (`<th>`) to specify the scope of the header. It helps assistive technologies understand the relationship between header cells and data cells, improving accessibility for users with disabilities. The `scope` attribute can take the following values:
    - `col`: Indicates that the header cell applies to the entire column.
    - `row`: Indicates that the header cell applies to the entire row.
    - `colgroup`: Indicates that the header cell applies to a group of columns defined by a `<colgroup>`.
    - `rowgroup`: Indicates that the header cell applies to a group of rows defined by a `<tbody>` or `<tfoot>`.

- Example of using the `scope` attribute:
    ```html
    <table>
        <tr>
            <th scope="col">Name</th>
            <th scope="col">Age</th>
            <th scope="col">City</th>
        </tr>
        <tr>
            <td>Alice</td>
            <td>30</td>
            <td>New York</td>
        </tr>
    </table>
    ```
    - In this example, the `scope="col"` attribute indicates that each header cell applies to its respective column, helping assistive technologies understand the structure of the table and improving accessibility for users with disabilities.

#### Id and Headers Attributes

- An alternative to using the `scope` attribute for associating header cells with data cells is to use the `id` and `headers` attributes. 

- The `id` attribute is assigned to header cells (`<th>`) to uniquely identify them, while the 

- `headers` attribute is used in data cells (`<td>`) to reference the corresponding header cells by their `id`. This method allows for more complex associations between header and data cells, especially in cases where a single data cell may be associated with multiple headers.

- Example: 
    ```html
    <table>
        <tr>
            <th id="name-header">Name</th>
            <th id="age-header">Age</th>
            <th id="city-header">City</th>
        </tr>
        <tr>
            <td headers="name-header">Alice</td>
            <td headers="age-header">30</td>
            <td headers="city-header">New York</td>
        </tr>
    </table>
    ```
    - In this example, the `id` attributes are assigned to the header cells, and the `headers` attributes in the data cells reference those `id`s to establish a clear association between the headers and their corresponding data cells. This method is particularly useful for complex tables where a single data cell may be associated with multiple headers.
