### SVG

#### Intro

- **_SVG_** stands for **Scalable Vector Graphics**. It is a markup language for describing two-dimensional graphics applications and images, and a set of related graphics script interfaces. SVG is used to define vector-based graphics for the Web, and it can be styled with CSS and scripted with JavaScript.

- In shorter terms: SVGs are a scalable image format, which means they will easily scale to any size and retain their quality without increasing their filesize. They’re also very useful if you need to create or modify your images programmatically, because you can change their properties through CSS and JavaScript.

- SVGs are defined using **_XML (Extensible Markup Language)_**, which means they can be created and edited with any text editor, as well as with drawing software. They can also be animated and interactive, making them a powerful tool for web design and development.

#### Anatomy of an SVG

- Example syntax:
    ```html
    <svg xmlns="http://www.w3.org/2000/svg" viewBox="0 0 100 100">
        <rect x=0 y=0 width=100 height=50 />
        <circle class="svg-circle" cx="50" cy="50" r="10"/>
    </svg>
    ```

1. `xlmns` attribute: This specifies the XML namespace for the SVG document. It is required for the SVG to be properly recognized and rendered by browsers. In simpler words, it tells the browser that this is an SVG document and how to interpret it.

2. `viewBox` attribute: This defines the coordinate system and dimensions of the SVG canvas. It takes four values: `min-x`, `min-y`, `width`, and `height`. In this example, it sets the origin of the coordinate system at (0, 0) and defines a canvas that is 100 units wide and 100 units tall.

3. `<rect>` element: This creates a rectangle shape in the SVG. The `x` and `y` attributes specify the position of the top-left corner of the rectangle, while the `width` and `height` attributes define its dimensions.

4. `<circle>` element: This creates a circle shape in the SVG. The `cx` and `cy` attributes specify the center of the circle, while the `r` attribute defines its radius. The `class` attribute can be used to apply CSS styles to the circle, allowing for further customization of its appearance.

- Find more on: https://developer.mozilla.org/en-US/docs/Web/SVG/Reference/Element 

#### Embedding SVGs

- There are 2 main ways to embed SVGs in your HTML:
    1. **Inline SVG:** You can include the SVG code directly within your HTML document, as shown in the example above. This allows for easy styling and scripting of the SVG elements using CSS and JavaScript. Exmple:
        ```html
            <svg xmlns="http://www.w3.org/2000/svg" viewBox="0 0 100 100">
                <rect x=0 y=0 width=100 height=50 />
                <circle class="svg-circle" cx="50" cy="50" r="10"/>
            </svg>
        ```

    2. **External/Linking SVG:** You can reference an external SVG file using the `<img>` tag or the `<object>` tag. This approach is useful when you want to reuse the same SVG in multiple places or when the SVG is complex and large. Example:
        ```html
            <img src="path/to/your/image.svg" alt="Description of the image">
        ```

### Tables

#### Intro

##### Basics

- Tables are a way to display data in a structured format, using rows and columns. They are created using the `<table>` element in HTML, along with various other elements to define the structure and content of the table.

- To create a table, you typically use the following elements:
    - `<table>`: This is the container element for the entire table.

    - `<th>`: This stands for **"table header"** and is used to define a header cell in the table. It is typically used in the first row of the table to label the columns.

    - `<tr>`: This stands for **"table row"** and is used to define a row in the table.

    - `<td>`: This stands for **"table data"** and is used to define a standard cell in the table that contains data.

- Example of a simple table:
    ```html
    <table>
        <tr>
            <th>Name</th>
            <th>Age</th>
            <th>City</th>
        </tr>
        <tr>
            <td>Alice</td>
            <td>30</td>
            <td>New York</td>
        </tr>
        <tr>
            <td>Bob</td>
            <td>25</td>
            <td>Los Angeles</td>
        </tr>
    </table>
    ```

##### Advanced

- `<thead>`: This element is used to group the header content in a table. It typically contains one or more `<tr>` elements that define the header rows of the table.

- `<tbody>`: This element is used to group the body content in a table. It typically contains one or more `<tr>` elements that define the body rows of the table.

- `<tfoot>`: This element is used to group the footer content in a table. It typically contains one or more `<tr>` elements that define the footer rows of the table.

- Example of a table with `<thead>`, `<tbody>`, and `<tfoot>`:
    ```html
    <table>
        <thead>
            <tr>
                <th>Name</th>
                <th>Age</th>
                <th>City</th>
            </tr>
        </thead>
        <tbody>
            <tr>
                <td>Alice</td>
                <td>30</td>
                <td>New York</td>
            </tr>
            <tr>
                <td>Bob</td>
                <td>25</td>
                <td>Los Angeles</td>
            </tr>
        </tbody>
        <tfoot>
            <tr>
                <td colspan="3">End of Table</td>
            </tr>
        </tfoot>
    </table>
    ```

#### Spanning Cells Across Multiple Rows & Columns

- `colspan` attribute: This is used to make a cell span across multiple columns. You can specify the number of columns you want the cell to span by setting the value of `colspan`. For example, if you want a cell to span across 3 columns, you would use `colspan="3"`. Example:
    ```html
    <table>
        <tr>
            <td colspan="2">Alice</td>
            <td>New York</td>
        </tr>
    </table>
    ```

- `rowspan` attribute: This is used to make a cell span across multiple rows. You can specify the number of rows you want the cell to span by setting the value of `rowspan`. For example, if you want a cell to span across 2 rows, you would use `rowspan="2"`. Example:
    ```html
    <table>
        <tr>
            <td rowspan="2">Alice</td>
            <td>30</td>
            <td>New York</td>
        </tr>
    </table>
    ```

#### Grouping Columns

- `<colgroup>` element: This is used to group columns in a table. It allows you to apply styles or attributes to a group of columns at once. You can specify the number of columns in the group using the `span` attribute. Example:
    ```html
    <table>
        <colgroup>
            <col span="2" style="background-color: lightgray;">
            <col style="background-color: lightblue;">
        </colgroup>
        <tr>
            <th>Name</th>
            <th>Age</th>
            <th>City</th>
        </tr>
        <tr>
            <td>Alice</td>
            <td>30</td>
            <td>New York</td>
        </tr>
    </table>
    ```
    - In this example, the first two columns `(Name and Age)` are grouped together and styled with a light gray background, while the third column `(City)` is styled with a light blue background.



- `<col>` element: This is used within a `<colgroup>` to define individual columns. You can apply styles or attributes to specific columns using the `<col>` element. Example:
    ```html
    <table>
        <colgroup>
            <col style="background-color: lightgray;">
            <col style="background-color: lightblue;">
            <col style="background-color: lightgreen;">
        </colgroup>
        <tr>
            <th>Name</th>
            <th>Age</th>
            <th>City</th>
        </tr>
        <tr>
            <td>Alice</td>
            <td>30</td>
            <td>New York</td>
        </tr>
    </table>
    ```
    - In this example, each column is defined with a different background color using the `<col>` element within the `<colgroup>`.

#### Captioning Tables

- `<caption>` element: This is used to provide a caption or title for the table. It is typically placed immediately after the opening `<table>` tag. The content of the `<caption>` element is displayed above the table by default, but it can be styled with CSS to change its position and appearance. Example:
    ```html
    <table>
        <caption>Employee Information</caption>
        <tr>
            <th>Name</th>
            <th>Age</th>
            <th>City</th>
        </tr>
        <tr>
            <td>Alice</td>
            <td>30</td>
            <td>New York</td>
        </tr>
    </table>
    ```
    - In this example, the caption "Employee Information" is displayed above the table, providing context for the data contained within the table.

#### Scope Attribute

- The `scope` attribute is used in table header cells (`<th>`) to specify the scope of the header. It helps assistive technologies understand the relationship between header cells and data cells, improving accessibility for users with disabilities. The `scope` attribute can take the following values:
    - `col`: Indicates that the header cell applies to the entire column.
    - `row`: Indicates that the header cell applies to the entire row.
    - `colgroup`: Indicates that the header cell applies to a group of columns defined by a `<colgroup>`.
    - `rowgroup`: Indicates that the header cell applies to a group of rows defined by a `<tbody>` or `<tfoot>`.

- Example of using the `scope` attribute:
    ```html
    <table>
        <tr>
            <th scope="col">Name</th>
            <th scope="col">Age</th>
            <th scope="col">City</th>
        </tr>
        <tr>
            <td>Alice</td>
            <td>30</td>
            <td>New York</td>
        </tr>
    </table>
    ```
    - In this example, the `scope="col"` attribute indicates that each header cell applies to its respective column, helping assistive technologies understand the structure of the table and improving accessibility for users with disabilities.

#### Id and Headers Attributes

- An alternative to using the `scope` attribute for associating header cells with data cells is to use the `id` and `headers` attributes. 

- The `id` attribute is assigned to header cells (`<th>`) to uniquely identify them, while the 

- `headers` attribute is used in data cells (`<td>`) to reference the corresponding header cells by their `id`. This method allows for more complex associations between header and data cells, especially in cases where a single data cell may be associated with multiple headers.

- Example: 
    ```html
    <table>
        <tr>
            <th id="name-header">Name</th>
            <th id="age-header">Age</th>
            <th id="city-header">City</th>
        </tr>
        <tr>
            <td headers="name-header">Alice</td>
            <td headers="age-header">30</td>
            <td headers="city-header">New York</td>
        </tr>
    </table>
    ```
    - In this example, the `id` attributes are assigned to the header cells, and the `headers` attributes in the data cells reference those `id`s to establish a clear association between the headers and their corresponding data cells. This method is particularly useful for complex tables where a single data cell may be associated with multiple headers.

# CSS

### CSS Units

- **_Absolute Units:_** The units that don't change based on the context of the element. They are fixed and will always be the same size regardless of the screen size or resolution. Examples include:
    - `px` (pixels): A pixel is a unit of measurement that represents a single point on a screen. It is the most commonly used absolute unit in CSS.
    
    - `cm` (centimeters): A centimeter is a unit of measurement that is equal to 10 millimeters. It is not commonly used in web design, but it can be useful for print stylesheets.

    - `mm` (millimeters): A millimeter is a unit of measurement that is equal to 1/10th of a centimeter. Like centimeters, it is not commonly used in web design.

    - `in` (inches): An inch is a unit of measurement that is equal to 2.54 centimeters. It is also not commonly used in web design.

- **_Relative Units:_** The units that change based on the context of the element. They are flexible and can adapt to different screen sizes and resolutions. Examples include:
    - `em`: This unit is relative to the font size of the parent element. For example, if the parent element has a font size of 16px, then 1em would be equal to 16px. If you set a child element's font size to 2em, it would be 32px (2 times the parent font size).

    - `rem`: This unit is relative to the root element's font size (usually the `<html>` element). For example, if the root element has a font size of 16px, then 1rem would be equal to 16px regardless of the font size of its parent elements.

    - `%` (percentage): This unit is relative to the size of the parent element. For example, if you set an element's width to 50%, it will take up half the width of its parent element.

    - `vw` (viewport width): This unit is relative to the width of the viewport (the visible area of the browser window). For example, if you set an element's width to 50vw, it will take up half the width of the viewport.

    - `vh` (viewport height): This unit is relative to the height of the viewport. For example, if you set an element's height to 50vh, it will take up half the height of the viewport.

### Advanced Selectors

#### Chaining Selectors

- Chaining selectors is a technique to select elements as a chain without separation. It allows you to target elements that have multiple classes or attributes. For example, if you want to select an element that has both the class "button" and the class "primary", you can use the following selector:
    ```html
        <div>
            <div class="subsection header">Latest Posts</div>
            <p class="subsection preview">This is where a preview for a post might go.</p>
        </div>
    ```
- In this example, the selector `.subsection.header` will target the `<div>` element that has both the "subsection" and "header" classes, while the selector `.subsection.preview` will target the `<p>` element that has both the "subsection" and "preview" classes. This technique allows for more specific targeting of elements based on their class combinations.

#### Descendant Combinators

##### Intro

- Descendant Combinators are used to select elements that are descendants of a specified element. The descendant combinator is represented by a space between two selectors.
- Example:
    ```html
        <div class="ancestor">
            <div class="contents">
                <div class="contents"></div>
            </div>
        </div>

        <div class="contents"></div>
    ```
- In this example, the selector `.ancestor .contents` will target both `<div>` elements with the class "contents" that are descendants of the `<div>` with the class "ancestor". However, it will not target the `<div>` with the class "contents" that is outside of the "ancestor" div. This allows you to select elements based on their hierarchical relationship in the DOM.


- There are 4 types of descendant combinators:
    1. **Child Combinator (`>`):** This combinator selects only the direct children of an element. For example, `div > p` will select all `<p>` elements that are direct children of a `<div>`.

    2. **Adjacent Sibling Combinator (`+`):** This combinator selects the element that is immediately following another element. For example, `div + p` will select all `<p>` elements that immediately follow a `<div>`.

    3. **General Sibling Combinator (`~`):** This combinator selects all elements that are siblings of a specified element, regardless of their position. For example, `div ~ p` will select all `<p>` elements that are siblings of a `<div>`, even if they are not immediately following it.

    4. **Descendant Combinator (space):** This combinator selects all elements that are descendants of a specified element, regardless of their level in the hierarchy. For example, `div p` will select all `<p>` elements that are descendants of a `<div>`, including those that are nested within other elements.

##### Child and Sibling Combinators

- Child and Sibling Combinators: These selectors allow you to target elements based on their relationship to other elements in the DOM. They include:
    1. `>` _(Child Combinator):_ This selector targets direct child elements of a specified parent. For example, `div > p` will select all `<p>` elements that are direct children of a `<div>`.

    2. `+` _(Adjacent Sibling Combinator):_ This selector targets the element that immediately follows another element. For example, `div + p` will select all `<p>` elements that immediately follow a `<div>`.

    3. `~` _(General Sibling Combinator):_ This selector targets all elements that follow another element. For example, `div ~ p` will select all `<p>` elements that follow a `<div>`.

- The difference between the adjacent sibling combinator (`+`) and the general sibling combinator (`~`) is that the adjacent sibling combinator only selects the immediate next sibling, while the general sibling combinator selects all siblings that follow the specified element, regardless of their position in relation to it.

#### Pseudo-selectors

##### Intro

- **_Pseudo-classes:_** selectors prefixed with a single colon (`:`) and target elements that already exist in HTML. They are used to define a special state of an element or to target elements based on their position in the document tree. Examples include `:hover`, `:first-child`, and `:nth-child()`.
- In simpler terms, pseudo-classes allow you to select **_elements_** based on their state or position in the document. 

- **_Pseudo-elements:_** selectors prefixed with a double colon (`::`) and target elements that don't normally exist in the markup. They are used to style specific parts of an element, such as the first letter or the first line of text. Examples include `::first-letter`, `::first-line`, and `::before`/`::after`.
- In simpler terms, pseudo-elements allow you to create and style **_parts of elements_** that don't exist in the HTML, giving you more control over the presentation of your content.

##### Pseudo-Classes

###### Dynamic and User Action Pseudo-classes

1. `:hover`: This pseudo-class applies when the user hovers over an element with a pointing device, such as a mouse. It is commonly used to change the appearance of buttons or links when they are hovered over.

2. `:focus`: This pseudo-class applies when an element has received focus, either through keyboard navigation or by being clicked on. It is often used to style form elements or interactive components when they are focused.

3. `:active`: This pseudo-class applies when an element is being activated by the user, such as when a button is being clicked. It is used to style elements in their active state.

4. `:link`: This pseudo-class applies to links that have not yet been visited by the user. It is used to style unvisited links.

5. `:visited`: This pseudo-class applies to links that have already been visited by the user. It is used to style visited links differently from unvisited links.

###### Structural Pseudo-classes

- **_Structural Pseudo-Classes:_** are a powerful way to select elements based on their position within the DOM.

1. `:root`: This pseudo-class selects the root element of the document, which is typically the `<html>` element. It is often used to define global CSS variables that can be accessed throughout the stylesheet.

2. `:first-child`: This pseudo-class selects an element that is the first child of its parent. For example, `p:first-child` will select any `<p>` element that is the first child of its parent.

3. `:last-child`: This pseudo-class selects an element that is the last child of its parent. For example, `p:last-child` will select any `<p>` element that is the last child of its parent.

4. `:nth-child(n)`: This pseudo-class selects an element that is the nth child of its parent, where n can be a number, a keyword (like `odd` or `even`), or a formula (like `2n+1`). For example, `p:nth-child(2)` will select the second `<p>` element that is a child of its parent, while `p:nth-child(odd)` will select all odd-numbered `<p>` elements.

5. `:empty`: This pseudo-class selects elements that have no children, including text nodes. For example, `div:empty` will select all `<div>` elements that do not contain any child elements or text content.

6. `:only-child`: This pseudo-class selects an element that is the only child of its parent. For example, `p:only-child` will select any `<p>` element that is the only child of its parent.

##### Pseudo-elements

- Pseudo-elements are more abstract. They allow us to affect parts of our HTML that aren’t elements at all.

1. `::marker`: This pseudo-element targets the marker box of a _list_ item, which is typically the bullet or number that appears before the content of a list item. It allows you to style the marker separately from the content of the list item. For example, you can change the color, size, or shape of the marker using this pseudo-element

2. `::first-letter`: This pseudo-element targets the first letter of a block-level element, such as a paragraph. It allows you to style the first letter differently from the rest of the text, which can be useful for creating drop caps or emphasizing the beginning of a section.

3. `::first-line`: This pseudo-element targets the first line of a block-level element. It allows you to style the first line differently from the rest of the text, which can be useful for creating emphasis or a unique design for the first line of a paragraph.

4. `::before` and `::after`: These pseudo-elements allow you to insert content before or after the content of an element. They are often used to add decorative elements or to create complex layouts without needing additional HTML markup. For example, you can use `::before` to add an icon before a heading, or `::after` to add a decorative flourish after a paragraph.

5. `::selection`: This pseudo-element targets the portion of an element that has been highlighted by the user, such as when they select text with their mouse. It allows you to style the selected text, such as changing its background color or text color.

#### Attribute Selectors

- Attribute selectors allow you to select elements based on the presence or value of their attributes. They are useful for targeting specific elements without needing to add additional classes or IDs to your HTML. Here are some common attribute selectors:

    1. `[attr]`: This selector targets elements that have a specific attribute, regardless of its value. For example, `[disabled]` will select all elements that have the `disabled` attribute.

    2. `selector[attr]`: This selector targets elements that have a specific attribute, regardless of its value. For example, `input[required]` will select all `<input>` elements that have the `required` attribute.

    3. `[attr="value"]`: This selector targets elements that have a specific attribute with a specific value. For example, `[href="https://www.example.com"]` will select all elements that have the `href` attribute with the value `https://www.example.com`.

    4. `[attr^="value"]`: This selector targets elements that have a specific attribute whose value starts with a specific string. For example, `[class^="btn"]` will select all elements that have the `class` attribute whose value starts with `btn`.

    5. `[attr$="value"]`: This selector targets elements that have a specific attribute whose value ends with a specific string. For example, `[src$=".jpg"]` will select all elements that have the `src` attribute whose value ends with `.jpg`.

    6. `[attr*="value"]`: This selector targets elements that have a specific attribute whose value contains a specific string. For example, `[alt*="image"]` will select all elements that have the `alt` attribute whose value contains the string `image`.

    7. `selector[attr|="value"]`: This selector targets elements that have a specific attribute whose value is either exactly "value" or starts with "value-" (where "value" is followed by a hyphen). For example, `[lang|="en"]` will select all elements that have the `lang` attribute with a value of "en" or any value that starts with "en-", such as "en-US" or "en-GB".

### Positioning

#### Static and Relative Positioning

- **_Static Positioning:_** This is the default positioning for all elements. Elements with static positioning are positioned according to the normal flow of the document. They are not affected by `top`, `bottom`, `left`, or `right` properties. Static elements will stack on top of each other based on their order in the HTML.

- **_Relative Positioning:_** This allows you to position an element relative to its normal position in the document flow. When you set an element's position to `relative`, it remains in the normal flow of the document, but you can use the `top`, `bottom`, `left`, and `right` properties to move it from its original position. The space originally occupied by the element is still preserved, so other elements will not fill that space. For example, if you set `top: 10px` on a relatively positioned element, it will move 10 pixels down from its original position, but the space it originally occupied will still be there. Example:

    ```html
        <div style="position: relative; top: 10px; left: 20px;">
            This element is relatively positioned.
        </div>
    ```

#### Absolute and Fixed Positioning

- Absolute Positioning: This allows you to position an element relative to its nearest positioned ancestor (an ancestor with a position of `relative`, `absolute`, or `fixed`). If there is no such ancestor, it will be positioned relative to the initial containing block (usually the `<html>` element). When you set an element's position to `absolute`, it is removed from the normal flow of the document, and other elements will fill the space it would have occupied. You can use the `top`, `bottom`, `left`, and `right` properties to specify its exact position. Example:

    ```html
        <div style="position: relative;">
            <div style="position: absolute; top: 10px; left: 20px;">
                This element is absolutely positioned.
            </div>
        </div>
    ```

    - Exampples of good use cases:
        1. Modals
        2. Tooltips
        3. Images with captions
        4. Icons on top of other elements
        
- NB: absolute positioning has very specific use cases and if possible, using flexbox or grid should be prioritized. Absolute positioning shouldn’t be used to do entire page layouts.

- Fixed Positioning: This allows you to position an element relative to the viewport, which means it will stay in the same position even when the page is scrolled. When you set an element's position to `fixed`, it is removed from the normal flow of the document, and other elements will fill the space it would have occupied. You can use the `top`, `bottom`, `left`, and `right` properties to specify its exact position. Example:

    ```html
        <div style="position: fixed; top: 0; left: 0; width: 100%; background-color: lightgray;">
            This element is fixed at the top of the viewport.
        </div>
    ```

    - Examples of good use cases:
        1. Navigation bars
        2. Back-to-top buttons
        3. Sticky headers
        4. Floating action buttons

#### Sticky Positioning

- Sticky Positioning: This allows you to position an element based on the user's scroll position. An element with `position: sticky` will behave like a relatively positioned element until it reaches a specified offset from the viewport (defined by `top`, `bottom`, `left`, or `right`), at which point it will "stick" to that position and behave like a fixed element. Once the user scrolls past the sticky element, it will return to its normal flow in the document. Example:

    ```html
        <div style="position: sticky; top: 0; background-color: lightgray;">
            This element will stick to the top of the viewport when you scroll.
        </div>
    ```

    - Examples of good use cases:
        1. Sticky headers
        2. Table of contents
        3. Sidebar navigation
        4. Call-to-action buttons

### CSS Functions

#### Intro

- CSS Functions are built-in functions that can be used in CSS property values to perform calculations, manipulate colors, or create dynamic effects. They allow you to create more complex and flexible styles without needing to write additional JavaScript. Here are some common CSS functions:

    1. `rgb()`: This function is used to define colors using the RGB color model. It takes three parameters representing the red, green, and blue components of the color, each ranging from 0 to 255. For example, `color: rgb(255, 0, 0);` will set the text color to red.

    2. `background: linear-gradient(90deg, red, blue);`: This function creates a linear gradient background that transitions from red to blue at a 90-degree angle. The `deg` parameter specifies the direction of the gradient, and the color parameters define the colors used in the gradient.

#### calc() Function

- The most powerful use cases for `calc()` are:
    1. Mixing Units: You can use `calc()` to mix different units in a single calculation. For example, `width: calc(100% - 50px);` allows you to set an element's width to be the full width of its container minus 50 pixels.

    2. The ability to nest: You can use `calc()` within another `calc()`, allowing for more complex calculations. For example, `calc(calc() - calc())` can be used to perform multiple calculations in a single property value. Example:
        ```css
            :root {
                --header: 3rem;
                --footer: 40px;
                --main: calc(100vh - calc(var(--header) + var(--footer)));
            }
        ```

#### min() and max() Functions

- `min()`: This function takes one or more comma-separated values and returns the smallest value. It is useful for setting a property to a value that should not exceed a certain limit. For example, `width: min(50%, 300px);` will set the width to 50% of the container, but it will not exceed 300 pixels.

- `max()`: This function takes one or more comma-separated values and returns the largest value. It is useful for setting a property to a value that should not be smaller than a certain limit. For example, `width: max(200px, 10%);` will set the width to 200 pixels, but it will not be smaller than 10% of the container.

#### clamp() Function

- `clamp()`: This function takes three parameters: a minimum value, a preferred value, and a maximum value. It returns the preferred value if it falls within the range defined by the minimum and maximum values; otherwise, it returns the closest value within that range. For example, `font-size: clamp(1rem, 2vw, 3rem);` will set the font size to 2% of the viewport width, but it will not be smaller than 1rem or larger than 3rem. This is particularly useful for creating responsive typography that adapts to different screen sizes while maintaining readability.

### Custom Properties (CSS Variables)

#### Intro

- The syntax for defining a custom property (CSS variable) is as follows:
    ```css
    :root {
        --main-color: #3498db;
        --secondary-color: #2ecc71;
        --font-size: 16px;
    }
    ```

- In this example, we define three custom properties: `--main-color`, `--secondary-color`, and `--font-size`. The `:root` selector is used to define these variables at the global level, making them accessible throughout the entire stylesheet.

- They are accessed using the `var()` function, which allows you to use the value of a custom property in any CSS property. For example:
    ```css
    body {
        background-color: var(--main-color);
        color: var(--secondary-color);
        font-size: var(--font-size);
    }
    ```

##### Fallback Values

- The `var()` functions accepts two parameters: the name of the custom property and an optional fallback value. The syntax is as follows:
    ```css
        body{
            color: var(--text-color, black);
        }
    ```

#### :root Selector

- `:root` is a pseudo-class selector that targets the root element of the document, which is typically the `<html>` element. It is commonly used to define global custom properties (CSS variables) that can be accessed throughout the entire stylesheet. By defining variables in the `:root` selector, you ensure that they are available for use in any other selector within your CSS. This allows for consistent styling and easy maintenance, as you can change the value of a variable in one place and have it update across all instances where it is used.

- Example:
    ```css
    :root {
        --primary-color: #3498db;
        --secondary-color: #2ecc71;
        --font-size: 16px;
    }

    body {
        background-color: var(--primary-color);
        color: var(--secondary-color);
        font-size: var(--font-size);
    }
    ```

- The `:root` selector is particularly useful for defining theme colors, font sizes, spacing values, and other design tokens that you want to reuse throughout your stylesheet. By centralizing these values in the `:root` selector, you can easily manage and update your design system without having to search through your CSS for individual property values.

- For example, to use in theme switching:
    ```css
        :root.dark-theme {
            --primary-color: #2c3e50;
            --secondary-color: #ecf0f1;
            --font-size: 16px;
        }

        :root.light-theme {
            --primary-color: #3498db;
            --secondary-color: #2ecc71;
            --font-size: 16px;
        }

        body {
            background-color: var(--primary-color);
            color: var(--secondary-color);
            font-size: var(--font-size);
        }
    ```

#### Media Queries

- The `prefers-color-scheme` media query is used to apply different styles based on the user's preferred color scheme, which can be either light or dark. This media query allows you to create a responsive design that adapts to the user's system settings, providing a better user experience.

- The syntax for using the `prefers-color-scheme` media query is as follows:

    ```css
    :root {
        --primary-color: #3498db;
        --secondary-color: #2ecc71;
        --font-size: 16px;
    }

            --primary-color: #3498db;
            --secondary-color: #2ecc71;
            --font-size: 16px;
        }
    }
    ```

    - In the example above, we define two media queries: one for users who prefer a dark color scheme and another for users who prefer a light color scheme. Inside each media query, we set different values for the custom properties defined in the `:root` selector. This allows the styles to automatically adjust based on the user's system preferences, providing a seamless and personalized experience.

    - First add the custom `root` properties outside of the media queries, then override them inside the media queries. This way, you ensure that there are default values for the custom properties, and they will be overridden based on the user's color scheme preference.

## Forms

### Basics

- The form element accepts _two essential attributes_:
    1. `action`: This attribute specifies the URL where the form data will be sent when the form is submitted. It can be a relative or absolute URL. For example, `action="/submit-form"` will send the form data to the `/submit-form` endpoint on the same server. In simpler terms, the `action` attribute tells the browser where to send the form data when the user submits the form.

    2. `method`: This attribute specifies the _HTTP method_ to be used when submitting the form. The most common methods are `GET` and `POST`. In simpler terms, the `method` attribute tells the browser how to send the form data to the server. 
    
        - `GET`: This method appends the form data to the URL as query parameters. It is typically used for retrieving data from the server. For example, if you have a form with an input field named "search" and the user enters "CSS", the form data will be sent as `?search=CSS` in the URL. In simpler terms, the `GET` method sends the form data in the URL, making it visible to the user and suitable for non-sensitive data retrieval.

        - `POST`: This method sends the form data in the body of the HTTP request. It is typically used for submitting data to the server, such as when creating a new resource or updating existing data. The form data is not visible in the URL when using `POST`. In simpler terms, the `POST` method sends the form data in a way that is not visible to the user, making it more suitable for sensitive data submission or when you want to send larger amounts of data.

- The basic syntax for a form element is as follows:
    ```html
    <form action="example.com/submit" method="POST">
        <!-- Form fields go here -->
    </form>
    ```

### Form Controls

#### Form Controls

- **_Form controls:_** These are the interactive elements that allow users to input data into a form. They include:

    1. `<input>`: This is a versatile element that can be used to create various types of form controls, such as text fields, checkboxes, radio buttons, and more. The type of control is determined by the `type` attribute. For example, `<input type="text">` creates a text field, while `<input type="checkbox">` creates a checkbox.

    - The `<input>` elements goes with the `<label>` element, which is used to provide a label for the input field. The `for` attribute of the `<label>` element should match the `id` of the corresponding `<input>` element to create an association between them. For example:
        ```html
        <label for="username">Username:</label>
        <input type="text" id="username" name="username">
        ```

    2. `<textarea>`: This element is used to create a multi-line text input field. It allows users to enter larger amounts of text compared to a single-line `<input>` element. For example, `<textarea name="message" rows="4" cols="50"></textarea>` creates a textarea with 4 rows and 50 columns.

    3. `<select>`: This element is used to create a dropdown list of options. It contains one or more `<option>` elements that define the available choices. For example:

        ```html
            <option value="usa">United States</option>
        ```

#### Form Control Attributes

- Form controls have attributes that define their behavior and appearance. Some common attributes include:

    1. `name`: This attribute specifies the name of the form control, which is used to identify the data when the form is submitted. For example, `<input type="text" name="username">` assigns the name "username" to the input field. It also assists the backend to know which data is being sent when the form is submitted. In simpler terms, the `name` attribute gives a name to the form control, allowing the server to recognize and process the data correctly when the form is submitted.

    2. `value`: This attribute specifies the initial value of the form control. For example, `<input type="text" name="username" value="JohnDoe">` sets the initial value of the input field to "JohnDoe".

    3. `placeholder`: This attribute provides a hint to the user about what kind of input is expected in the form control. It displays a short description inside the input field when it is empty. For example, `<input type="text" name="username" placeholder="Enter your username">` shows "Enter your username" as a placeholder in the input field.

    4. `required`: This boolean attribute indicates that a form control must be filled out before submitting the form. If a required field is left empty, the browser will display a validation message and prevent form submission. For example, `<input type="email" name="email" required>` makes the email input field mandatory.

    5. `disabled`: This boolean attribute indicates that a form control is disabled and cannot be interacted with by the user. Disabled controls are typically grayed out and do not submit their values when the form is submitted. For example, `<input type="text" name="username" disabled>` creates a text input field that is disabled.

    6. `for`: This attribute is used with the `<label>` element to associate it with a specific form control. The value of the `for` attribute should match the `id` of the corresponding form control.             <input type="email" id="email" name="email">

### Form Validation

#### Validation Attributes

- Form Validation is the process of ensuring that the data entered into a form meets certain criteria before it is submitted to the server. This can be done using HTML5 built-in validation attributes, which provide a simple way to validate form inputs without needing to write custom JavaScript. Some common validation attributes include:

    1. `required`: This attribute indicates that a form control must be filled out before submitting the form. If a required field is left empty, the browser will display a validation message and prevent form submission. For example, `<input type="email" name="email" required>` makes the email input field mandatory.

    2. `type`: This attribute specifies the type of form control and can also trigger built-in validation based on the type. For example, `<input type="email" name="email">` will validate that the input is in a valid email format.

    3. `pattern`: This attribute allows you to specify a regular expression that the input value must match in order for the form to be submitted. For example, `<input type="text" name="username" pattern="[A-Za-z0-9]{5,}">` requires the username to be at least 5 characters long and contain only letters and numbers.

    4. `min` and `max`: These attributes are used with numeric input types (like `number`, `range`, `date`, etc.) to specify the minimum and maximum values allowed. For example, `<input type="number" name="age" min="18" max="99">` restricts the age input to values between 18 and 99.

    5. `minlength` and `maxlength`: These attributes specify the minimum and maximum number of characters allowed in a text input field. For example, `<input type="text" name="username" minlength="5" maxlength="15">` requires the username to be between 5 and 15 characters long.

#### Styling Validation States

1. `:user-valid`: This pseudo-class applies to form controls that have been validated by the browser and are considered valid according to the specified validation rules. For example, if you have an input field with the `required` attribute and the user has entered a value, the input will be considered valid, and you can style it using `:user-valid`.

2. `:user-invalid`: This pseudo-class applies to form controls that have been validated by the browser and are considered invalid according to the specified validation rules. For example, if you have an input field with the `type="email"` attribute and the user has entered an invalid email address, the input will be considered invalid, and you can style it using `:user-invalid`.

    ```css
            border-radius: 5px;
        }

        input:user-invalid {
            border-color: red;
        }

        input:user-valid {
            border-color: green;
        }
    ```

## Grid

### Basics

#### Creating Grids

- To create a grid layout, you can use the `display: grid;` property on a container element. This will turn the container into a grid and allow you to define rows and columns for your layout. For example:
    ```css
    .grid-container {
        display: grid;
        grid-template-columns: 1fr 1fr 1fr; /* Creates three equal columns */
        grid-template-rows: auto; /* Rows will adjust based on content */
        gap: 10px; /* Adds space between grid items */
    }
    ```

#### Lines & Tracks

- **_Lines:_** In CSS Grid, lines are _the invisible boundaries that define the start and end of rows and columns._ They are numbered starting from 1 at the top-left corner of the grid. 
- You can refer to these lines when placing grid items using properties like `grid-column` and `grid-row`. 
- For example, `grid-column: 1 / 3;` will make a grid item span from line 1 to line 3, effectively covering two columns.

- **_Tracks:_** Tracks are _the actual rows and columns in a grid._ They are defined by the `grid-template-columns` and `grid-template-rows` properties. 
- Each track can have a specific size, which can be defined using various units such as pixels, percentages, or the `fr` unit (which represents a fraction of the available space). 
- For example, `grid-template-columns: 100px 2fr 1fr;` creates three columns where the first column is 100 pixels wide, the second column takes up twice as much space as the third column, and the third column takes up one fraction of the remaining space. In this case, the second column will be twice as wide as the third column, and the first column will have a fixed width of 100 pixels.

#### Columns & Rows

- **_Columns:_** Columns in CSS Grid are defined using the `grid-template-columns` property. This property allows you to specify the width of each column in the grid. You can use various units to define column widths, such as pixels (`px`), percentages (`%`), or the `fr` unit (which represents a fraction of the available space). For example:
    ```css
    .grid-container {
        display: grid;
        grid-template-columns: 100px 2fr 1fr; /* Creates three columns with specified widths */
    }
    ```

- **_Rows:_** Rows in CSS Grid are defined using the `grid-template-rows` property. This property allows you to specify the height of each row in the grid. Similar to columns, you can use various units to define row heights, such as pixels (`px`), percentages (`%`), or the `fr` unit (which represents a fraction of the available space). For example:
    ```css
    .grid-container {
        display: grid;
        grid-template-rows: 50px auto 100px; /* Creates three rows with specified heights */
    }
    ```

- Grid-Template Shorthand: You can also use the `grid-template` shorthand property to define both rows and columns in a single declaration. The syntax for the `grid-template` shorthand is as follows:
    ```css
    .grid-container {
        display: grid;
        grid-template: 
            "header header header" 50px
            "sidebar main main" auto
            "footer footer footer" 100px;
    }
    ```

    - In this example, we define a grid with three rows and three columns. The first row is 50 pixels tall and contains the "header" area spanning all three columns. The second row is set to `auto`, allowing it to adjust based on the content, and contains the "sidebar" area in the first column and the "main" area spanning the second and third columns. The third row is 100 pixels tall and contains the "footer" area spanning all three columns.

#### Explicit & Implicit Grids

- **_Explicit Grid:_** An explicit grid is a grid that is defined using the `grid-template-columns` and `grid-template-rows` properties. It specifies the number of rows and columns in the grid, as well as their sizes. For example:
    ```css
    .grid-container {
        display: grid;
        grid-template-columns: 100px 200px 1fr; /* Explicitly defines three columns */
        grid-template-rows: 50px 100px; /* Explicitly defines two rows */
    }
    ```

- **_Implicit Grid:_** An implicit grid is created automatically by the browser when you place grid items outside the bounds of the explicit grid. If you place a grid item in a position that exceeds the defined rows or columns, the browser will create additional rows or columns to accommodate the item. For example:
    ```css
    .grid-container {
        display: grid;
        grid-template-columns: 100px 200px; /* Explicitly defines two columns */
        grid-template-rows: 50px; /* Explicitly defines one row */
    }

    .grid-item {
        grid-column: 3; /* This will create an implicit third column */
        grid-row: 2; /* This will create an implicit second row */
    }
    ```

    - In this example, we have defined an explicit grid with two columns and one row. However, when we place a grid item in the third column and second row, the browser automatically creates an implicit third column and second row to accommodate the item. This allows for flexibility in layout design, as you can add items without needing to predefine every possible row and column in your grid.

#### Gap

- The gap between grid items can be controlled using the `gap` property, which is a shorthand for `row-gap` and `column-gap`. The `gap` property allows you to specify the size of the space between rows and columns in a grid layout. For example:
    ```css
    .grid-container {
        display: grid;
        grid-template-columns: 1fr 1fr 1fr; /* Creates three equal columns */
        grid-template-rows: auto; /* Rows will adjust based on content */
        gap: 20px; /* Adds a 20px gap between all rows and columns */
    }
    ```

### Advanced

#### Repeat

- The `repeat()` function in CSS Grid is a convenient way to define repeated patterns of rows or columns without having to write out each one individually. It allows you to specify a pattern and the number of times it should be repeated. The syntax for the `repeat()` function is as follows:
    ```css
    .grid-container {
        display: grid;
        grid-template-columns: repeat(3, 1fr); /* Creates three equal and flexible columns */
        grid-template-rows: repeat(2, 50px); /* Creates two rows, each fixed 50px tall */
    }
    ```

    - In this example, `repeat(3, 1fr)` creates three columns that each take up an equal fraction of the available space (1fr), while `repeat(2, 50px)` creates two rows that are each 50 pixels tall. The `repeat()` function simplifies the process of defining multiple rows or columns with the same size or pattern, making your CSS more concise and easier to read.

- We can also use different values in the `repeat()` function to create more complex patterns. For example:
    ```css
    .grid-container {
        display: grid;
        grid-template-columns: repeat(2, 100px) 1fr; /* Creates two fixed columns and one flexible column */
        grid-template-rows: repeat(3, 50px); /* Creates three rows, each fixed 50px tall */
    }
    ```

    - In this example, `repeat(2, 100px)` creates two columns that are each 100 pixels wide, followed by a third column that takes up the remaining space (1fr). The rows are defined using `repeat(3, 50px)`, which creates three rows that are each 50 pixels tall. This demonstrates how the `repeat()` function can be used to create a variety of grid layouts with different patterns and sizes.

- Another example using just the `fr` units:
    ```css
    .grid-container {
        display: grid;
        grid-template-columns: repeat(2, 2fr) repeat(3, 1fr); /* Creates two columns that are twice as wide as the next three columns */
        grid-template-rows: repeat(4, 1fr); /* Creates four equal rows */
    }
    ```
    - In this example, `repeat(2, 2fr)` creates two columns that are each twice as wide as the next three columns created by `repeat(3, 1fr)`. The rows are defined using `repeat(4, 1fr)`, which creates four rows that each take up an equal fraction of the available space. This illustrates how the `repeat()` function can be used to create complex grid layouts with varying column widths and row heights.
