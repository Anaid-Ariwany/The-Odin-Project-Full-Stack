### SVG

#### Intro

- **_SVG_** stands for **Scalable Vector Graphics**. It is a markup language for describing two-dimensional graphics applications and images, and a set of related graphics script interfaces. SVG is used to define vector-based graphics for the Web, and it can be styled with CSS and scripted with JavaScript.

- In shorter terms: SVGs are a scalable image format, which means they will easily scale to any size and retain their quality without increasing their filesize. They’re also very useful if you need to create or modify your images programmatically, because you can change their properties through CSS and JavaScript.

- SVGs are defined using **_XML (Extensible Markup Language)_**, which means they can be created and edited with any text editor, as well as with drawing software. They can also be animated and interactive, making them a powerful tool for web design and development.

#### Anatomy of an SVG

- Example syntax:
    ```html
    <svg xmlns="http://www.w3.org/2000/svg" viewBox="0 0 100 100">
        <rect x=0 y=0 width=100 height=50 />
        <circle class="svg-circle" cx="50" cy="50" r="10"/>
    </svg>
    ```

1. `xlmns` attribute: This specifies the XML namespace for the SVG document. It is required for the SVG to be properly recognized and rendered by browsers. In simpler words, it tells the browser that this is an SVG document and how to interpret it.

2. `viewBox` attribute: This defines the coordinate system and dimensions of the SVG canvas. It takes four values: `min-x`, `min-y`, `width`, and `height`. In this example, it sets the origin of the coordinate system at (0, 0) and defines a canvas that is 100 units wide and 100 units tall.

3. `<rect>` element: This creates a rectangle shape in the SVG. The `x` and `y` attributes specify the position of the top-left corner of the rectangle, while the `width` and `height` attributes define its dimensions.

4. `<circle>` element: This creates a circle shape in the SVG. The `cx` and `cy` attributes specify the center of the circle, while the `r` attribute defines its radius. The `class` attribute can be used to apply CSS styles to the circle, allowing for further customization of its appearance.

- Find more on: https://developer.mozilla.org/en-US/docs/Web/SVG/Reference/Element 

#### Embedding SVGs

- There are 2 main ways to embed SVGs in your HTML:
    1. **Inline SVG:** You can include the SVG code directly within your HTML document, as shown in the example above. This allows for easy styling and scripting of the SVG elements using CSS and JavaScript. Exmple:
        ```html
            <svg xmlns="http://www.w3.org/2000/svg" viewBox="0 0 100 100">
                <rect x=0 y=0 width=100 height=50 />
                <circle class="svg-circle" cx="50" cy="50" r="10"/>
            </svg>
        ```

    2. **External/Linking SVG:** You can reference an external SVG file using the `<img>` tag or the `<object>` tag. This approach is useful when you want to reuse the same SVG in multiple places or when the SVG is complex and large. Example:
        ```html
            <img src="path/to/your/image.svg" alt="Description of the image">
        ```

### Tables

#### Intro

##### Basics

- Tables are a way to display data in a structured format, using rows and columns. They are created using the `<table>` element in HTML, along with various other elements to define the structure and content of the table.

- To create a table, you typically use the following elements:
    - `<table>`: This is the container element for the entire table.

    - `<th>`: This stands for **"table header"** and is used to define a header cell in the table. It is typically used in the first row of the table to label the columns.

    - `<tr>`: This stands for **"table row"** and is used to define a row in the table.

    - `<td>`: This stands for **"table data"** and is used to define a standard cell in the table that contains data.

- Example of a simple table:
    ```html
    <table>
        <tr>
            <th>Name</th>
            <th>Age</th>
            <th>City</th>
        </tr>
        <tr>
            <td>Alice</td>
            <td>30</td>
            <td>New York</td>
        </tr>
        <tr>
            <td>Bob</td>
            <td>25</td>
            <td>Los Angeles</td>
        </tr>
    </table>
    ```

##### Advanced

- `<thead>`: This element is used to group the header content in a table. It typically contains one or more `<tr>` elements that define the header rows of the table.

- `<tbody>`: This element is used to group the body content in a table. It typically contains one or more `<tr>` elements that define the body rows of the table.

- `<tfoot>`: This element is used to group the footer content in a table. It typically contains one or more `<tr>` elements that define the footer rows of the table.

- Example of a table with `<thead>`, `<tbody>`, and `<tfoot>`:
    ```html
    <table>
        <thead>
            <tr>
                <th>Name</th>
                <th>Age</th>
                <th>City</th>
            </tr>
        </thead>
        <tbody>
            <tr>
                <td>Alice</td>
                <td>30</td>
                <td>New York</td>
            </tr>
            <tr>
                <td>Bob</td>
                <td>25</td>
                <td>Los Angeles</td>
            </tr>
        </tbody>
        <tfoot>
            <tr>
                <td colspan="3">End of Table</td>
            </tr>
        </tfoot>
    </table>
    ```

#### Spanning Cells Across Multiple Rows & Columns

- `colspan` attribute: This is used to make a cell span across multiple columns. You can specify the number of columns you want the cell to span by setting the value of `colspan`. For example, if you want a cell to span across 3 columns, you would use `colspan="3"`. Example:
    ```html
    <table>
        <tr>
            <td colspan="2">Alice</td>
            <td>New York</td>
        </tr>
    </table>
    ```

- `rowspan` attribute: This is used to make a cell span across multiple rows. You can specify the number of rows you want the cell to span by setting the value of `rowspan`. For example, if you want a cell to span across 2 rows, you would use `rowspan="2"`. Example:
    ```html
    <table>
        <tr>
            <td rowspan="2">Alice</td>
            <td>30</td>
            <td>New York</td>
        </tr>
    </table>
    ```

#### Grouping Columns

- `<colgroup>` element: This is used to group columns in a table. It allows you to apply styles or attributes to a group of columns at once. You can specify the number of columns in the group using the `span` attribute. Example:
    ```html
    <table>
        <colgroup>
            <col span="2" style="background-color: lightgray;">
            <col style="background-color: lightblue;">
        </colgroup>
        <tr>
            <th>Name</th>
            <th>Age</th>
            <th>City</th>
        </tr>
        <tr>
            <td>Alice</td>
            <td>30</td>
            <td>New York</td>
        </tr>
    </table>
    ```
    - In this example, the first two columns `(Name and Age)` are grouped together and styled with a light gray background, while the third column `(City)` is styled with a light blue background.



- `<col>` element: This is used within a `<colgroup>` to define individual columns. You can apply styles or attributes to specific columns using the `<col>` element. Example:
    ```html
    <table>
        <colgroup>
            <col style="background-color: lightgray;">
            <col style="background-color: lightblue;">
            <col style="background-color: lightgreen;">
        </colgroup>
        <tr>
            <th>Name</th>
            <th>Age</th>
            <th>City</th>
        </tr>
        <tr>
            <td>Alice</td>
            <td>30</td>
            <td>New York</td>
        </tr>
    </table>
    ```
    - In this example, each column is defined with a different background color using the `<col>` element within the `<colgroup>`.

#### Captioning Tables

- `<caption>` element: This is used to provide a caption or title for the table. It is typically placed immediately after the opening `<table>` tag. The content of the `<caption>` element is displayed above the table by default, but it can be styled with CSS to change its position and appearance. Example:
    ```html
    <table>
        <caption>Employee Information</caption>
        <tr>
            <th>Name</th>
            <th>Age</th>
            <th>City</th>
        </tr>
        <tr>
            <td>Alice</td>
            <td>30</td>
            <td>New York</td>
        </tr>
    </table>
    ```
    - In this example, the caption "Employee Information" is displayed above the table, providing context for the data contained within the table.

#### Scope Attribute

- The `scope` attribute is used in table header cells (`<th>`) to specify the scope of the header. It helps assistive technologies understand the relationship between header cells and data cells, improving accessibility for users with disabilities. The `scope` attribute can take the following values:
    - `col`: Indicates that the header cell applies to the entire column.
    - `row`: Indicates that the header cell applies to the entire row.
    - `colgroup`: Indicates that the header cell applies to a group of columns defined by a `<colgroup>`.
    - `rowgroup`: Indicates that the header cell applies to a group of rows defined by a `<tbody>` or `<tfoot>`.

- Example of using the `scope` attribute:
    ```html
    <table>
        <tr>
            <th scope="col">Name</th>
            <th scope="col">Age</th>
            <th scope="col">City</th>
        </tr>
        <tr>
            <td>Alice</td>
            <td>30</td>
            <td>New York</td>
        </tr>
    </table>
    ```
    - In this example, the `scope="col"` attribute indicates that each header cell applies to its respective column, helping assistive technologies understand the structure of the table and improving accessibility for users with disabilities.

#### Id and Headers Attributes

- An alternative to using the `scope` attribute for associating header cells with data cells is to use the `id` and `headers` attributes. 

- The `id` attribute is assigned to header cells (`<th>`) to uniquely identify them, while the 

- `headers` attribute is used in data cells (`<td>`) to reference the corresponding header cells by their `id`. This method allows for more complex associations between header and data cells, especially in cases where a single data cell may be associated with multiple headers.

- Example: 
    ```html
    <table>
        <tr>
            <th id="name-header">Name</th>
            <th id="age-header">Age</th>
            <th id="city-header">City</th>
        </tr>
        <tr>
            <td headers="name-header">Alice</td>
            <td headers="age-header">30</td>
            <td headers="city-header">New York</td>
        </tr>
    </table>
    ```
    - In this example, the `id` attributes are assigned to the header cells, and the `headers` attributes in the data cells reference those `id`s to establish a clear association between the headers and their corresponding data cells. This method is particularly useful for complex tables where a single data cell may be associated with multiple headers.

### CSS Units

- **_Absolute Units:_** The units that don't change based on the context of the element. They are fixed and will always be the same size regardless of the screen size or resolution. Examples include:
    - `px` (pixels): A pixel is a unit of measurement that represents a single point on a screen. It is the most commonly used absolute unit in CSS.
    
    - `cm` (centimeters): A centimeter is a unit of measurement that is equal to 10 millimeters. It is not commonly used in web design, but it can be useful for print stylesheets.

    - `mm` (millimeters): A millimeter is a unit of measurement that is equal to 1/10th of a centimeter. Like centimeters, it is not commonly used in web design.

    - `in` (inches): An inch is a unit of measurement that is equal to 2.54 centimeters. It is also not commonly used in web design.

- **_Relative Units:_** The units that change based on the context of the element. They are flexible and can adapt to different screen sizes and resolutions. Examples include:
    - `em`: This unit is relative to the font size of the parent element. For example, if the parent element has a font size of 16px, then 1em would be equal to 16px. If you set a child element's font size to 2em, it would be 32px (2 times the parent font size).

    - `rem`: This unit is relative to the root element's font size (usually the `<html>` element). For example, if the root element has a font size of 16px, then 1rem would be equal to 16px regardless of the font size of its parent elements.

    - `%` (percentage): This unit is relative to the size of the parent element. For example, if you set an element's width to 50%, it will take up half the width of its parent element.

    - `vw` (viewport width): This unit is relative to the width of the viewport (the visible area of the browser window). For example, if you set an element's width to 50vw, it will take up half the width of the viewport.

    - `vh` (viewport height): This unit is relative to the height of the viewport. For example, if you set an element's height to 50vh, it will take up half the height of the viewport.

### Advanced Selectors

#### Chaining Selectors

- Chaining selectors is a technique to select elements as a chain without separation. It allows you to target elements that have multiple classes or attributes. For example, if you want to select an element that has both the class "button" and the class "primary", you can use the following selector:
    ```html
        <div>
            <div class="subsection header">Latest Posts</div>
            <p class="subsection preview">This is where a preview for a post might go.</p>
        </div>
    ```
- In this example, the selector `.subsection.header` will target the `<div>` element that has both the "subsection" and "header" classes, while the selector `.subsection.preview` will target the `<p>` element that has both the "subsection" and "preview" classes. This technique allows for more specific targeting of elements based on their class combinations.

#### Descendant Combinators

##### Intro

- Descendant Combinators are used to select elements that are descendants of a specified element. The descendant combinator is represented by a space between two selectors.
- Example:
    ```html
        <div class="ancestor">
            <div class="contents">
                <div class="contents"></div>
            </div>
        </div>

        <div class="contents"></div>
    ```
- In this example, the selector `.ancestor .contents` will target both `<div>` elements with the class "contents" that are descendants of the `<div>` with the class "ancestor". However, it will not target the `<div>` with the class "contents" that is outside of the "ancestor" div. This allows you to select elements based on their hierarchical relationship in the DOM.


- There are 4 types of descendant combinators:
    1. **Child Combinator (`>`):** This combinator selects only the direct children of an element. For example, `div > p` will select all `<p>` elements that are direct children of a `<div>`.

    2. **Adjacent Sibling Combinator (`+`):** This combinator selects the element that is immediately following another element. For example, `div + p` will select all `<p>` elements that immediately follow a `<div>`.

    3. **General Sibling Combinator (`~`):** This combinator selects all elements that are siblings of a specified element, regardless of their position. For example, `div ~ p` will select all `<p>` elements that are siblings of a `<div>`, even if they are not immediately following it.

    4. **Descendant Combinator (space):** This combinator selects all elements that are descendants of a specified element, regardless of their level in the hierarchy. For example, `div p` will select all `<p>` elements that are descendants of a `<div>`, including those that are nested within other elements.

##### Child and Sibling Combinators

- Child and Sibling Combinators: These selectors allow you to target elements based on their relationship to other elements in the DOM. They include:
    1. `>` _(Child Combinator):_ This selector targets direct child elements of a specified parent. For example, `div > p` will select all `<p>` elements that are direct children of a `<div>`.

    2. `+` _(Adjacent Sibling Combinator):_ This selector targets the element that immediately follows another element. For example, `div + p` will select all `<p>` elements that immediately follow a `<div>`.

    3. `~` _(General Sibling Combinator):_ This selector targets all elements that follow another element. For example, `div ~ p` will select all `<p>` elements that follow a `<div>`.

- The difference between the adjacent sibling combinator (`+`) and the general sibling combinator (`~`) is that the adjacent sibling combinator only selects the immediate next sibling, while the general sibling combinator selects all siblings that follow the specified element, regardless of their position in relation to it.

#### Pseudo-selectors

##### Intro

- **_Pseudo-classes:_** selectors prefixed with a single colon (`:`) and target elements that already exist in HTML. They are used to define a special state of an element or to target elements based on their position in the document tree. Examples include `:hover`, `:first-child`, and `:nth-child()`.
- In simpler terms, pseudo-classes allow you to select **_elements_** based on their state or position in the document. 

- **_Pseudo-elements:_** selectors prefixed with a double colon (`::`) and target elements that don't normally exist in the markup. They are used to style specific parts of an element, such as the first letter or the first line of text. Examples include `::first-letter`, `::first-line`, and `::before`/`::after`.
- In simpler terms, pseudo-elements allow you to create and style **_parts of elements_** that don't exist in the HTML, giving you more control over the presentation of your content.

##### Pseudo-Classes

###### Dynamic and User Action Pseudo-classes

1. `:hover`: This pseudo-class applies when the user hovers over an element with a pointing device, such as a mouse. It is commonly used to change the appearance of buttons or links when they are hovered over.

2. `:focus`: This pseudo-class applies when an element has received focus, either through keyboard navigation or by being clicked on. It is often used to style form elements or interactive components when they are focused.

3. `:active`: This pseudo-class applies when an element is being activated by the user, such as when a button is being clicked. It is used to style elements in their active state.

4. `:link`: This pseudo-class applies to links that have not yet been visited by the user. It is used to style unvisited links.

5. `:visited`: This pseudo-class applies to links that have already been visited by the user. It is used to style visited links differently from unvisited links.

###### Structural Pseudo-classes

- **_Structural Pseudo-Classes:_** are a powerful way to select elements based on their position within the DOM.

1. `:root`: This pseudo-class selects the root element of the document, which is typically the `<html>` element. It is often used to define global CSS variables that can be accessed throughout the stylesheet.

2. `:first-child`: This pseudo-class selects an element that is the first child of its parent. For example, `p:first-child` will select any `<p>` element that is the first child of its parent.

3. `:last-child`: This pseudo-class selects an element that is the last child of its parent. For example, `p:last-child` will select any `<p>` element that is the last child of its parent.

4. `:nth-child(n)`: This pseudo-class selects an element that is the nth child of its parent, where n can be a number, a keyword (like `odd` or `even`), or a formula (like `2n+1`). For example, `p:nth-child(2)` will select the second `<p>` element that is a child of its parent, while `p:nth-child(odd)` will select all odd-numbered `<p>` elements.

5. `:empty`: This pseudo-class selects elements that have no children, including text nodes. For example, `div:empty` will select all `<div>` elements that do not contain any child elements or text content.

6. `:only-child`: This pseudo-class selects an element that is the only child of its parent. For example, `p:only-child` will select any `<p>` element that is the only child of its parent.

##### Pseudo-elements

- Pseudo-elements are more abstract. They allow us to affect parts of our HTML that aren’t elements at all.

1. `::marker`: This pseudo-element targets the marker box of a _list_ item, which is typically the bullet or number that appears before the content of a list item. It allows you to style the marker separately from the content of the list item. For example, you can change the color, size, or shape of the marker using this pseudo-element

2. `::first-letter`: This pseudo-element targets the first letter of a block-level element, such as a paragraph. It allows you to style the first letter differently from the rest of the text, which can be useful for creating drop caps or emphasizing the beginning of a section.

3. `::first-line`: This pseudo-element targets the first line of a block-level element. It allows you to style the first line differently from the rest of the text, which can be useful for creating emphasis or a unique design for the first line of a paragraph.

4. `::before` and `::after`: These pseudo-elements allow you to insert content before or after the content of an element. They are often used to add decorative elements or to create complex layouts without needing additional HTML markup. For example, you can use `::before` to add an icon before a heading, or `::after` to add a decorative flourish after a paragraph.

5. `::selection`: This pseudo-element targets the portion of an element that has been highlighted by the user, such as when they select text with their mouse. It allows you to style the selected text, such as changing its background color or text color.
